# Visualization for negative edges

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
import matplotlib.patches as mpatches
import os

# ==========================================
# 1. 配置与定义
# ==========================================

NODE_BIT_MAP = {
    'is_kinase': 0,
    'is_disease_related': 1
}

EDGE_BIT_MAP = {
    "is_negative": 0,           # Bit 0
    "location_match_y": 1,      # Bit 1
    "M_sim": 2,                 # Bit 2
    "physical_interaction": 3,  # Bit 3
    "x_hub_degree": 4,          # Bit 4
    "x_low_betweenness": 5,     # Bit 5
    "Affinity_Capture_MS": 6,   # Bit 6
}

NODE_LABELS = {
    0: "Other",
    1: "Kinase",
    2: "Disease",
    3: "Kinase+Disease"
}

NODE_COLOR_LIST = ['#d3d3d3', '#1f77b4', '#ff7f0e', '#d62728']

# ==========================================
# 2. 规则解码器 (支持开关切换)
# ==========================================

def decode_gar_rules(X, E, mode='active_only'):
    """
    mode: 
      - 'active_only': 只显示成立(Bit=1)的谓词 (除了基础拓扑 l/¬l 总是显示)
      - 'full_dual': 显示所有谓词的 True/False 状态 (用于调试)
    """
    literals = []
    num_nodes = len(X)
    
    # --- A. 节点谓词 ---
    for i in range(num_nodes):
        val = X[i]
        props = []
        if (val >> NODE_BIT_MAP['is_kinase']) & 1: props.append("Kinase")
        if (val >> NODE_BIT_MAP['is_disease_related']) & 1: props.append("Disease")
        
        # active_only 模式下，如果节点没有任何属性，就不显示该节点的描述
        if mode == 'active_only':
            if props:
                literals.append(f"n{i}: {{{','.join(props)}}}")
        else:
            # full_dual 模式下，没有属性也显示 Other
            if not props: props.append("Other")
            literals.append(f"n{i}: {{{','.join(props)}}}")

    # --- B. 边谓词 ---
    rows, cols = E.shape
    for u in range(rows):
        for v in range(u + 1, cols):
            raw_val = E[u][v]
            if raw_val == 0: continue # 无边
            
            feat_mask = raw_val - 1
            current_edge_preds = []
            
            # 1. 基础拓扑 (无论什么模式都必须显示，因为这是边的存在性)
            is_neg = (feat_mask >> EDGE_BIT_MAP["is_negative"]) & 1
            if is_neg:
                base = f"¬l(n{u},n{v})" # 负边
            else:
                base = f"l(n{u},n{v})"  # 正边
            
            # 2. 遍历其他特征
            # 定义特征对应的 Logic 描述 (Key -> (True描述, False描述))
            logic_map = {
                "location_match_y":     ("n{u}.Loc==n{v}.Loc",  "n{u}.Loc≠n{v}.Loc"),
                "M_sim":                ("M(n{u},n{v})",        "¬M(n{u},n{v})"),
                "physical_interaction": ("Type==Phys",          "Type≠Phys"),
                "x_hub_degree":         ("Deg≥Hub",             "Deg<Hub"),
                "x_low_betweenness":    ("Betw≤Low",            "Betw>Low"),
                "Affinity_Capture_MS":  ("Sys==MS",             "Sys≠MS")
            }

            extras = []
            for key, (true_str, false_str) in logic_map.items():
                bit_val = (feat_mask >> EDGE_BIT_MAP[key]) & 1
                
                if mode == 'full_dual':
                    # 全显示模式：显示 True 或 False
                    extras.append(true_str if bit_val else false_str)
                elif mode == 'active_only':
                    # 仅显示模式：只显示 True
                    if bit_val:
                        extras.append(true_str)

            # 3. 组合输出
            # 格式: l(n1,n2) ∧ {Phys, Hub...}
            if extras:
                full_str = f"{base} ∧ {{{', '.join(extras)}}}"
            else:
                full_str = f"{base}"
            
            literals.append(full_str)

    return literals

# ==========================================
# 3. 文件解析器 (补全)
# ==========================================
def parse_graphs_from_txt(file_path):
    if not os.path.exists(file_path):
        print(f"Error: 文件 {file_path} 不存在")
        return []
    graphs = []
    current_x = None
    with open(file_path, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    i = 0
    while i < len(lines):
        line = lines[i].strip()
        if not line:
            i += 1; continue
        if line.startswith("X:"):
            if i + 1 < len(lines):
                try:
                    current_x = np.array([int(v) for v in lines[i+1].strip().split()])
                    i += 1
                except ValueError: pass
        elif line.startswith("E:") and current_x is not None:
            num_nodes = len(current_x)
            e_matrix = []
            valid = True
            for r in range(num_nodes):
                if i + 1 + r < len(lines):
                    try:
                        row = [int(v) for v in lines[i + 1 + r].strip().split()]
                        if len(row) == num_nodes: e_matrix.append(row)
                        else: valid = False
                    except: valid = False
                else: valid = False
            if valid and len(e_matrix) == num_nodes:
                graphs.append((current_x, np.array(e_matrix)))
                i += num_nodes; current_x = None
        i += 1
    return graphs

# ==========================================
# 4. 可视化函数 (带开关)
# ==========================================

def visualize_graph(x, e, graph_id=0, show_all_predicates=False):
    """
    show_all_predicates: True 则显示所有0/1状态，False 则只显示成立的特征
    """
    num_nodes = len(x)
    
    # 边提取
    pos_edges, neg_edges = [], []
    rows, cols = e.shape
    for i in range(rows):
        for j in range(i + 1, cols):
            raw_val = e[i][j]
            if raw_val == 0: continue
            feat_mask = raw_val - 1
            is_negative = (feat_mask >> EDGE_BIT_MAP["is_negative"]) & 1
            if is_negative: neg_edges.append((i, j))
            else: pos_edges.append((i, j))

    # 过滤空图 (可选)
    if len(neg_edges) == 0 and len(pos_edges) == 0: return 

    # ★★★ 调用解码器，传入模式 ★★★
    mode = 'full_dual' if show_all_predicates else 'active_only'
    literals = decode_gar_rules(x, e, mode=mode)
    
    # 控制台打印
    print(f"\n{'='*20} Graph #{graph_id} [{'FULL' if show_all_predicates else 'ACTIVE ONLY'}] {'='*20}")
    for p in literals:
        print(f"  {p}")
    print("="*50)

    # 绘图
    G = nx.Graph()
    node_colors = []
    categories_present = set()
    
    for i in range(num_nodes):
        feat_val = x[i] if x[i] <= 3 else 0
        G.add_node(i, category=NODE_LABELS[feat_val])
        categories_present.add(feat_val)
        node_colors.append(NODE_COLOR_LIST[feat_val])
        
    plt.figure(figsize=(14, 8))
    pos = nx.spring_layout(G, seed=42, k=0.9)
    
    nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=800, edgecolors='black')
    nx.draw_networkx_labels(G, pos, font_color='white', font_weight='bold')
    
    if pos_edges: nx.draw_networkx_edges(G, pos, edgelist=pos_edges, edge_color='black', width=2)
    if neg_edges: nx.draw_networkx_edges(G, pos, edgelist=neg_edges, edge_color='red', style='dashed', width=2.5)
    
    # Legend
    legend_handles = []
    for cat_idx in sorted(list(categories_present)):
        patch = mpatches.Patch(color=NODE_COLOR_LIST[cat_idx], label=f"{cat_idx}: {NODE_LABELS[cat_idx]}")
        legend_handles.append(patch)
    legend_handles.append(plt.Line2D([0], [0], color='black', lw=2, label='l(x,y)'))
    legend_handles.append(plt.Line2D([0], [0], color='red', lw=2, linestyle='--', label='¬l(x,y)'))

    plt.legend(handles=legend_handles, loc='upper left', bbox_to_anchor=(1.01, 1))

    # 文本框显示
    display_text = "Predicates:\n" + "\n".join(literals[:20])
    if len(literals) > 20:
        display_text += f"\n... (+{len(literals)-20} more)"
        
    plt.text(1.02, 0.5, display_text, transform=plt.gca().transAxes, 
             fontsize=10, verticalalignment='center', fontfamily='monospace',
             bbox=dict(boxstyle='round', facecolor='white', alpha=0.9, edgecolor='gray'))

    plt.title(f"Graph #{graph_id}\nMode: {mode}")
    plt.tight_layout()
    plt.show()

# ==========================================
# 5. 主程序
# ==========================================
# ==========================================
# 5. 主程序 (已修改：只显示含负边的图)
# ==========================================
# FILE_PATH = "/home/yyyy/codework/GARplus/DiGress/outputs/2026-01-23/11-15-51-ppi_gar/generated_samples1.txt"
FILE_PATH = "//home/yyyy/codework/GARplus/DiGress/outputs/2026-03-11/12-13-57-ppi_gar/generated_samples1.txt"
print(f"Reading from {FILE_PATH}...")
parsed_graphs = parse_graphs_from_txt(FILE_PATH)
print(f"Found {len(parsed_graphs)} graphs. Filtering for NEGATIVE edges...")

# ★★★ 开关配置 ★★★
SHOW_ALL = False  # False=只显示成立特征, True=显示所有特征(包括False的)

count_shown = 0

for idx, (x, e) in enumerate(parsed_graphs):
    # --- 过滤逻辑开始 ---
    has_negative_edge = False
    rows, cols = e.shape
    
    # 扫描上三角矩阵寻找负边
    for r in range(rows):
        for c in range(r + 1, cols):
            val = e[r][c]
            if val == 0: continue # 无边跳过
            
            # 还原 mask 并检查 bit 0
            feat_mask = val - 1
            if (feat_mask >> EDGE_BIT_MAP["is_negative"]) & 1:
                has_negative_edge = True
                break # 找到一条就可以停了
        if has_negative_edge:
            break
            
    # --- 过滤逻辑结束 ---

    # 只有包含负边的图才显示
    if has_negative_edge:
        visualize_graph(x, e, graph_id=idx, show_all_predicates=SHOW_ALL)
        count_shown += 1

print(f"\nDone. Displayed {count_shown} graphs with negative edges.")
```

# Support and confidence computation for PPI on training data

```python
import networkx as nx
import time
import signal
import matplotlib.pyplot as plt
from networkx.algorithms.isomorphism import GraphMatcher
import os
import traceback
import sys
from collections import Counter

# ================= 配置区 =================
MATCH_LIMIT =10000
TIME_LIMIT = 5

TARGET_FILE = "/home/yyyy/codework/GARplus/DiGress/outputs/2026-01-23/11-15-51-ppi_gar/generated_samples1.txt"
RAW_DATA_FILE = "/home/yyyy/codework/GARplus/DiGress/outputs/2026-01-23/11-15-51-ppi_gar/train_graphs_ground_truth.txt" 

# GAR+ 编码
NEGATIVE_LABELS = [3, 4]

# ★★★ 新增：开关配置 ★★★
ENABLE_NODE_MATCH = True  # True: 必须节点类型也相同; False: 只看拓扑结构
# =========================================

class TimeoutException(Exception): pass
def timeout_handler(signum, frame): raise TimeoutException()
signal.signal(signal.SIGALRM, timeout_handler)

# ----------------------------------------------
# 1. 匹配逻辑
# ----------------------------------------------
def edge_match(d1, d2):
    return d1.get('label') == d2.get('label')

def node_match_fn(n1, n2):
    """ 比较两个节点的类别 (X特征) """
    # 假设解析时把类别存为了 'label' 或 'cat'
    # 注意：networkx 默认用 'label' 做图同构的节点匹配吗？不是，需要显式指定函数
    return n1.get('cat') == n2.get('cat')

# ----------------------------------------------
# 2. 解析逻辑 (需增强以读取 X)
# ----------------------------------------------
def parse_graph_txt(filepath):
    """ 解析 TXT，返回 [(x_list, edge_matrix), ...] """
    graphs = []
    if not os.path.exists(filepath): return []
    
    with open(filepath, "r") as f: lines = f.readlines()
    i = 0
    while i < len(lines):
        line = lines[i].strip()
        if line.startswith("N="):
            try:
                n = int(line.split("=")[1])
                i += 1
                
                # 读取 X
                while i < len(lines) and not lines[i].strip().startswith("X:"): i += 1
                if i < len(lines): i += 1 # 跳过 'X:' 行
                
                x_list = []
                if i < len(lines):
                    x_str = lines[i].strip()
                    x_list = list(map(int, x_str.split()))
                    i += 1
                
                # 读取 E
                while i < len(lines) and not lines[i].strip().startswith("E:"): i += 1
                
                edge_matrix = []
                if i < len(lines):
                    i += 1
                    for _ in range(n):
                        if i >= len(lines): break
                        row_vals = list(map(int, lines[i].strip().replace(',', ' ').split()))
                        edge_matrix.append(row_vals)
                        i += 1
                        
                    if len(edge_matrix) == n and len(x_list) == n:
                        graphs.append((x_list, edge_matrix))
            except: i += 1
        else:
            i += 1
    return graphs

def to_nx_graph(x_list, edge_matrix):
    """ 将解析出的数据转为 NX Graph (包含节点属性) """
    n = len(edge_matrix)
    G = nx.Graph()
    
    # 添加节点及属性
    for i in range(n):
        G.add_node(i, cat=x_list[i]) # 存为 'cat' 属性
        
    # 添加边
    for i in range(n):
        for j in range(i + 1, n):
            val = edge_matrix[i][j]
            if val != 0: 
                G.add_edge(i, j, label=val)
    return G

def load_digress_as_big_graph_undirected(filepath):
    BigG = nx.Graph()
    if not os.path.exists(filepath):
        print(f"[Error] File not found: {filepath}")
        return BigG

    print(f"Parsing BigGraph from: {filepath} ...")
    graphs_data = parse_graph_txt(filepath)
    
    node_offset = 0
    total_edges = 0

    for x_list, mat in graphs_data:
        n = len(mat)
        
        # 添加节点 (带 Offset 和 Attribute)
        for i in range(n):
            BigG.add_node(i + node_offset, cat=x_list[i])

        # 添加边
        for i in range(n):
            for j in range(i + 1, n):
                val = mat[i][j]
                if val != 0:
                    u = i + node_offset
                    v = j + node_offset
                    if BigG.has_edge(u, v):
                        BigG[u][v]['label'] = max(BigG[u][v]['label'], val)
                    else:
                        BigG.add_edge(u, v, label=val)
                        total_edges += 1
        node_offset += n

    print(f"BigGraph Constructed. Nodes: {BigG.number_of_nodes()}, Edges: {BigG.number_of_edges()}")
    return BigG

# ----------------------------------------------
# 3. 计算逻辑
# ----------------------------------------------
def compute_structural_confidence(subG, bigG, match_limit, time_limit, use_node_match=False):
    edges_to_test = list(subG.edges(data=True))
    if not edges_to_test: return [] 

    results = []
    target_candidates = [
        (u, v) for u, v, d in edges_to_test 
        if d.get('label') in NEGATIVE_LABELS
    ]
    if not target_candidates:
        target_candidates = [(u, v) for u, v, d in edges_to_test][:3]

    # 根据开关决定是否传入 node_match 函数
    nm_func = node_match_fn if use_node_match else None

    for target_u, target_v in target_candidates:
        original_label = subG[target_u][target_v]['label']
        premiseG = subG.copy()
        premiseG.remove_edge(target_u, target_v)
        
        if premiseG.number_of_edges() == 0: continue

        GM = GraphMatcher(bigG, premiseG, node_match=nm_func, edge_match=edge_match)
        
        supp_premise = 0
        supp_negative = 0
        # 0: 无边, 1/2: 正边, 3/4: 负边
        label_dist = Counter()
        signal.alarm(time_limit)
        status = "Finished"
        
        try:
            for mapping in GM.subgraph_isomorphisms_iter():
                # --- 映射修复 ---
                valid_mapping = None
                if (target_u in mapping) and (target_v in mapping): valid_mapping = mapping
                else:
                    inv_map = {v: k for k, v in mapping.items()}
                    if (target_u in inv_map) and (target_v in inv_map): valid_mapping = inv_map
                
                if valid_mapping is None: continue
                # ----------------
                
                supp_premise += 1
                real_u = valid_mapping[target_u]
                real_v = valid_mapping[target_v]

                real_label = 0 # 默认为 0 (无边)
                if bigG.has_edge(real_u, real_v):
                    edge_data = bigG.get_edge_data(real_u, real_v)
                    real_label = edge_data.get('label', 0)
                
                # 更新计数器
                label_dist[real_label] += 1

                # 仍然统计 supp_negative 用于计算 Conf
                if real_label in NEGATIVE_LABELS:
                    supp_negative += 1
                
                if supp_premise >= match_limit:
                    status = "Limit"
                    break
            signal.alarm(0)

        except TimeoutException:
            status = "TimeOut"
        except Exception as e:
            signal.alarm(0)
            status = "Error"
            continue

        if supp_premise > 0:
            confidence = supp_negative / supp_premise
            # 将 Counter 转为排序后的字典字符串，方便查看
            dist_str = str(dict(sorted(label_dist.items())))
            
            results.append({
                'target_edge': (target_u, target_v),
                'original_label': original_label,
                'confidence': confidence,
                'support_negative': supp_negative,
                'support_shape': supp_premise,
                'label_dist': dist_str, # <--- 存入结果
                'status': status
            })
            
    return results

# ----------------------------------------------
# 4. 主程序
# ----------------------------------------------
# ----------------------------------------------
if __name__ == "__main__":
    print(f"Node Matching Enabled: {ENABLE_NODE_MATCH}")
    print(f"Strategy: Only check Largest Connected Component (LCC)")
    
    bigG = load_digress_as_big_graph_undirected(RAW_DATA_FILE)
    if bigG.number_of_nodes() == 0:
        print("EXITING: BigGraph is empty.")
        exit()

    graphs_data = parse_graph_txt(TARGET_FILE)
    
    # ★★★ 修改表头：增加 TruthDist 列，且留宽一点 ★★★
    print(f"\n{'ID':<4} | {'Type':<4} | {'Conf':<7} | {'SuppNeg':<7} | {'SuppShape':<9} | {'TruthDist (0=NoEdge)':<30} | {'Status'}")
    print("-" * 100)
    
    cnt_graphs_with_neg = 0
    
    for idx, (x_list, edge_mat) in enumerate(graphs_data):
        full_G = to_nx_graph(x_list, edge_mat)
        
        # 统计整图是否有负边 (为了计数)
        has_neg_anywhere = any(d.get('label') in NEGATIVE_LABELS for _,_,d in full_G.edges(data=True))
        if has_neg_anywhere:
            cnt_graphs_with_neg += 1
        
        # =========================================================
        # ★★★ 修改核心：提取最大连通分量 (LCC) ★★★
        # =========================================================
        if full_G.number_of_nodes() == 0:
            continue
            
        # 1. 获取所有连通分量 (generator -> list)
        comps = list(nx.connected_components(full_G))
        
        # 2. 找到包含节点数最多的那个集合
        largest_comp_nodes = max(comps, key=len)
        
        # 3. 创建子图
        lcc_G = full_G.subgraph(largest_comp_nodes).copy()
        
        # 4. 只有当 LCC 至少有 1 条边时才处理 (孤立点没有拓扑意义)
        if lcc_G.number_of_edges() < 1:
            continue
            
        # 5. 放入列表，只处理这一个
        valid_components = [lcc_G]
        # =========================================================

        for sub_idx, subG in enumerate(valid_components):
            # 检查 LCC 内部是否有负边
            has_negative_edge = False
            for u, v, d in subG.edges(data=True):
                if d.get('label') in NEGATIVE_LABELS:
                    has_negative_edge = True
                    break
            
            # 如果 LCC 里没有负边，就跳过 (哪怕其他小碎片里有，也不管了)
            if not has_negative_edge: continue
            
            metrics = compute_structural_confidence(
                subG, bigG, MATCH_LIMIT, TIME_LIMIT, 
                use_node_match=ENABLE_NODE_MATCH
            )
            
            if not metrics: continue
            
            best_m = max(metrics, key=lambda x: x['confidence'])
            print(f"{idx:<4} | {best_m['original_label']:<4} | {best_m['confidence']:.4f}  | {best_m['support_negative']:<7} | {best_m['support_shape']:<9} | {best_m['label_dist']:<30} | {best_m['status']}")
            
    print(f"\nTotal graphs containing negative edges (Anywhere): {cnt_graphs_with_neg}")
```

```python
import networkx as nx
import time
import signal
import matplotlib.pyplot as plt
from networkx.algorithms.isomorphism import GraphMatcher
import os
import traceback
import sys
from collections import Counter
import pandas as pd
import numpy as np

# ================= 配置区 =================
MATCH_LIMIT = 500
TIME_LIMIT = 5

# 输入文件
TARGET_FILE = "/home/yyyy/codework/GARplus/DiGress/outputs/2026-01-23/11-15-51-ppi_gar/generated_samples1.txt"

# 原始数据文件
RAW_EDGE_FILE = "/home/yyyy/codework/GARplus/DiGress/DiGress-main/data/PPI/raw/protein_protein_with_type.csv"
RAW_NODE_FILE = "/home/yyyy/codework/GARplus/DiGress/DiGress-main/data/PPI/raw/protein.csv"
UPDATE_EDGE_FILE = "/home/yyyy/codework/GARplus/DiGress/DiGress-main/data/PPI/raw/extended_update/protein_protein_with_type_update.csv"

# 阈值
ML_THRESHOLD = 0.5 

NODE_BIT_MAP = {'is_kinase': 0, 'is_disease_related': 1}
EDGE_BIT_MAP = {"is_negative": 0}

# 开关 (保持你现在的设置)
ENABLE_NODE_MATCH = True
ENABLE_PROPAGATION = True  
# =========================================

class TimeoutException(Exception): pass
def timeout_handler(signum, frame): raise TimeoutException()
signal.signal(signal.SIGALRM, timeout_handler)

# ----------------------------------------------
# 1. 核心编码逻辑
# 换了数据集以后主要需要修改的地方
# ----------------------------------------------
def encode_node_feature(row_series):
    n_feature = 0
    keywords = str(row_series.get('Keywords', ''))
    families = str(row_series.get('Protein families', ''))
    disease = row_series.get('Involvement in disease', None)
    
    if 'Kinase' in keywords or 'ATP-binding' in keywords or 'Kinase' in families:
        n_feature |= (1 << NODE_BIT_MAP['is_kinase'])
    if pd.notna(disease) and str(disease).strip() != '':
        n_feature |= (1 << NODE_BIT_MAP['is_disease_related'])
    return n_feature

def is_label_negative(val):
    if val == 0: return False
    mask = val - 1
    return (mask >> EDGE_BIT_MAP["is_negative"]) & 1

def get_gar_edge_index(edge_type_str, score):
    s = str(edge_type_str).lower()
    is_negative = 1 if 'negative' in s or 'added_negative' in s else 0
    return (is_negative << 0) + 1

# ----------------------------------------------
# 2. 匹配函数
# ----------------------------------------------
def node_match_fn(n1, n2):
    return n1.get('cat') == n2.get('cat')

#TODO
def edge_match_fn(d1, d2):
    # d1: BigGraph (1/2), d2: GenGraph (Bitmask)
    # 两者经过 is_label_negative 后应该是一致的
    l1 = d1.get('label', 0)
    l2 = d2.get('label', 0)
    return is_label_negative(l1) == is_label_negative(l2)

# ----------------------------------------------
# 3. 大图加载逻辑
# ----------------------------------------------
def load_csv_as_big_graph(edge_csv_path, node_csv_path):
    print(f"[Loader] Building BigGraph...")
    id_to_cat = {}
    try:
        df_meta = pd.read_csv(node_csv_path, low_memory=False)
        df_meta.columns = df_meta.columns.str.strip()
        id_col = 'BioGRID ID' 
        if id_col not in df_meta.columns: id_col = df_meta.columns[0]
        df_meta[id_col] = pd.to_numeric(df_meta[id_col], errors='coerce').fillna(-1).astype(int).astype(str)
        
        for _, row in df_meta.iterrows():
            if row[id_col] == '-1': continue
            id_to_cat[row[id_col]] = encode_node_feature(row)
        print(f"  - Metadata count: {len(id_to_cat)}")
    except Exception as e:
        print(f"  - [Error] Meta: {e}"); return nx.Graph()

    BigG = nx.Graph()
    try:
        df_ppi = pd.read_csv(edge_csv_path, sep=',')
        if len(df_ppi.columns) < 5: df_ppi = pd.read_csv(edge_csv_path, sep='\t')
        df_ppi.columns = df_ppi.columns.str.strip()
        
        col_ua = 'Official Symbol Interactor A'
        col_ub = 'Official Symbol Interactor B'
        col_id_a = 'BioGRID ID Interactor A'
        col_id_b = 'BioGRID ID Interactor B'
        
        neg_count = 0
        
        for _, row in df_ppi.iterrows():
            u_sym = str(row.get(col_ua, row[0])) 
            v_sym = str(row.get(col_ub, row[1]))
            u_bid = str(row.get(col_id_a, '')).split('.')[0]
            v_bid = str(row.get(col_id_b, '')).split('.')[0]
            
            u_cat = id_to_cat.get(u_bid, 0)
            v_cat = id_to_cat.get(v_bid, 0)
            
            edge_type = row.get('type', 'positive')
            score = row.get('Score', 0.0)
            gar_label = get_gar_edge_index(edge_type, score)
            
            if is_label_negative(gar_label): neg_count += 1
            
            if not BigG.has_node(u_sym): BigG.add_node(u_sym, cat=u_cat)
            if not BigG.has_node(v_sym): BigG.add_node(v_sym, cat=v_cat)
            
            if BigG.has_edge(u_sym, v_sym):
                BigG[u_sym][v_sym]['label'] = max(BigG[u_sym][v_sym]['label'], gar_label)
            else:
                BigG.add_edge(u_sym, v_sym, label=gar_label)
        
        print(f"  - BigGraph Negative Edges: {neg_count}")
                
    except Exception as e:
        print(f"  - [Error] Edges: {e}"); return nx.Graph()

    print(f"BigGraph Loaded. Nodes: {BigG.number_of_nodes()}, Edges: {BigG.number_of_edges()}")
    return BigG

# ----------------------------------------------
# 4. 解析与匹配 (带 DEBUG)
# ----------------------------------------------
def parse_graph_txt(filepath):
    # (省略部分解析代码，假设解析是正确的，这里不做改动)
    graphs = []
    if not os.path.exists(filepath): return []
    with open(filepath, "r") as f: lines = f.readlines()
    i = 0
    while i < len(lines):
        line = lines[i].strip()
        if line.startswith("N="):
            try:
                n = int(line.split("=")[1])
                i += 1
                while i < len(lines) and not lines[i].strip().startswith("X:"): i += 1
                if i < len(lines): i += 1 
                x_list = list(map(int, lines[i].strip().split())) if i < len(lines) else []
                i += 1
                while i < len(lines) and not lines[i].strip().startswith("E:"): i += 1
                edge_matrix = []
                if i < len(lines):
                    i += 1
                    for _ in range(n):
                        if i >= len(lines): break
                        edge_matrix.append(list(map(int, lines[i].strip().replace(',', ' ').split())))
                        i += 1
                    graphs.append((x_list, edge_matrix))
            except: i += 1
        else: i += 1
    return graphs

def to_nx_graph(x_list, edge_matrix):
    n = len(edge_matrix)
    G = nx.Graph()
    for i in range(n): G.add_node(i, cat=x_list[i]) 
    for i in range(n):
        for j in range(i+1, n):
            #TODO 
            if edge_matrix[i][j] != 0: G.add_edge(i, j, label=edge_matrix[i][j])
    return G

def compute_structural_confidence(subG, bigG, match_limit, time_limit, use_node_match=False):
    edges_to_test = list(subG.edges(data=True))
    
    # ★ DEBUG: 打印子图基本信息
    print(f"    [DEBUG] Checking subgraph: Nodes={subG.number_of_nodes()}, Edges={subG.number_of_edges()}")
    
    target_candidates = [(u, v) for u, v, d in edges_to_test if is_label_negative(d.get('label', 0))]
    
    # ★ DEBUG: 打印负边数量
    if not target_candidates:
        print("    [DEBUG] -> SKIPPED: No negative edges found in this subgraph.")
        return []
    else:
        print(f"    [DEBUG] -> Found {len(target_candidates)} negative edge(s) to verify.")

    nm_func = node_match_fn if use_node_match else None

    results = []
    for target_u, target_v in target_candidates:
        original_label = subG[target_u][target_v]['label']
        premiseG = subG.copy()
        premiseG.remove_edge(target_u, target_v)
        
        # ★ DEBUG: 检查前件图
        if premiseG.number_of_edges() == 0:
            # print(f"    [DEBUG] -> Edge ({target_u},{target_v}): Skipped because PremiseG is empty after removal.")
            continue

        # print(f"    [DEBUG] -> Matching PremiseG (Edges={premiseG.number_of_edges()}) for target ({target_u},{target_v})...")

        GM = GraphMatcher(bigG, premiseG, node_match=nm_func, edge_match=edge_match_fn)
        
        supp_premise = 0
        supp_negative = 0
        label_dist = Counter()
        no_edge_pairs = []
        # ★ DEBUG: 检查是否有同构
        # 我们先检查是否存在至少一个匹配，不进行完整迭代，以测试是否结构本身就不存在
        if not GM.subgraph_is_isomorphic():
             print("    [DEBUG] -> FAIL: No subgraph isomorphism found at all! (Structure does not exist in BigGraph)")
             continue
        
        # 如果能匹配到，开始迭代
        signal.alarm(time_limit)
        status = "Finished"
        try:
            for mapping in GM.subgraph_isomorphisms_iter():
                valid_mapping = None
                if (target_u in mapping) and (target_v in mapping): valid_mapping = mapping
                else:
                    inv_map = {v: k for k, v in mapping.items()}
                    if (target_u in inv_map) and (target_v in inv_map): valid_mapping = inv_map
                if valid_mapping is None: continue
                
                supp_premise += 1
                real_u = valid_mapping[target_u]
                real_v = valid_mapping[target_v]
                
                real_label = 0 
                if bigG.has_edge(real_u, real_v):
                    real_label = bigG[real_u][real_v].get('label', 0)
                
                label_dist[real_label] += 1
                if is_label_negative(real_label): supp_negative += 1
                #记录本来无边的到有边里面
                if real_label == 0: no_edge_pairs.append((real_u, real_v))
                
                # ★ DEBUG: 打印第一个匹配到的结果作为示例
                if supp_premise == 1:
                     print(f"    [DEBUG] -> First match found! Mapped to BigGraph nodes ({real_u}, {real_v}). Real Label: {real_label}")

                if supp_premise >= match_limit: break
            signal.alarm(0)
        except: 
            status = "TimeOut"
            print("    [DEBUG] -> TimeOut")

        print(f"    [DEBUG] -> Finished matching. Supp={supp_premise}, Neg={supp_negative}")

        if supp_premise > 0:
            confidence = supp_negative / supp_premise
            dist_str = str(dict(sorted(label_dist.items())))
            results.append({
                'target_edge': (target_u, target_v),
                'original_label': original_label,
                'confidence': confidence,
                'support_negative': supp_negative,
                'support_shape': supp_premise,
                'label_dist': dist_str,
                'no_edge_pairs': no_edge_pairs,
                'status': status
            })
    return results

# ==============================================================================
# 5. CSV 更新逻辑 (保持不变，省略)
# ==============================================================================
def apply_updates_to_csv_file(input_path, output_path, updates_list):
    """
    input_path: 原始 CSV
    output_path: 保存修改后的 CSV
    updates_list: list of (u, v)
    """
    if not updates_list: return
    
    print(f"\n[CSV Update] Reading original file: {input_path} ...")
    
    # 1. 读取原始 CSV
    try:
        df = pd.read_csv(input_path, sep=',')
        if len(df.columns) < 5: df = pd.read_csv(input_path, sep='\t')
    except Exception as e:
        print(f"Error reading CSV: {e}")
        return

    # 清理列名
    df.columns = df.columns.str.strip()
    col_a = 'Official Symbol Interactor A'
    col_b = 'Official Symbol Interactor B'
    col_type = 'type'
    
    # 确保有 type 列，没有则创建
    if col_type not in df.columns:
        df[col_type] = 'positive' 
    
    print(f"[CSV Update] Processing {len(updates_list)} updates...")
    
    updated_count = 0
    new_rows = [] # 用于收集新增的行
    
    # 2. 遍历每一个建议的负边
    for u, v in updates_list:
        # A. 检查是否存在 (u, v) 或 (v, u)
        mask = ((df[col_a] == u) & (df[col_b] == v)) | \
               ((df[col_a] == v) & (df[col_b] == u))
        
        indices = df.index[mask].tolist()
        
        if len(indices) > 0:
            # --- 场景 1: 行已存在 (在原 DataFrame 上直接修改) ---
            for idx in indices:
                old_val = df.at[idx, col_type]
                if old_val != 'added_negative':
                    print(f"  [MOD] Row {idx}: {u}-{v} | type: '{old_val}' -> 'added_negative'")
                    df.at[idx, col_type] = 'added_negative'
                    updated_count += 1
        else:
            # --- 场景 2: 行不存在 (收集到列表，稍后统一添加) ---
            print(f"  [ADD] New Row: {u}-{v} | type set to 'added_negative'")
            
            new_row = {col: np.nan for col in df.columns} # 创建空字典
            new_row[col_a] = u
            new_row[col_b] = v
            new_row[col_type] = 'added_negative'
            if 'Score' in df.columns: new_row['Score'] = 0.0
            
            new_rows.append(new_row)

    # 3. 将新增的行合并到 DataFrame
    if len(new_rows) > 0:
        print(f"[CSV Update] Appending {len(new_rows)} new rows...")
        df_new = pd.DataFrame(new_rows)
        df = pd.concat([df, df_new], ignore_index=True)

    # 4. 保存到新文件
    print(f"\n[CSV Update] Summary: Modified {updated_count} existing rows, Appended {len(new_rows)} new rows.")
    print(f"[CSV Update] Saving to NEW file: {output_path} ...")
    
    sep = '\t' if output_path.endswith('.tsv') else ','
    df.to_csv(output_path, index=False, sep=sep)
    print("Done! File saved successfully.")


# ----------------------------------------------
# 6. 主程序
# ----------------------------------------------
if __name__ == "__main__":
    bigG = load_csv_as_big_graph(RAW_EDGE_FILE, RAW_NODE_FILE)
    if bigG.number_of_nodes() == 0: exit()

    graphs_data = parse_graph_txt(TARGET_FILE)
    print(f"Loaded {len(graphs_data)} generated graphs.")
    
    print(f"\n{'ID':<4} | {'Type':<4} | {'Conf':<7} | {'SuppNeg':<7} | {'SuppShape':<9} | {'TruthDist':<20} | {'Status'}")
    print("-" * 90)
    
    # ★ DEBUG: 强制只跑前 5 个图，避免刷屏
    # count_run = 0
    all_updates = []
    for idx, (x_list, edge_mat) in enumerate(graphs_data):
        # if count_run >= 5: break # 只看前5个

        full_G = to_nx_graph(x_list, edge_mat)
        
        # 1. 检查节点数
        if full_G.number_of_nodes() == 0:
            print(f"[Graph {idx}] Skipped: Empty nodes")
            continue
            
        comps = list(nx.connected_components(full_G))
        if not comps: continue
        lcc_G = full_G.subgraph(max(comps, key=len)).copy()
        
        # 2. 检查边数
        if lcc_G.number_of_edges() < 1:
            print(f"[Graph {idx}] Skipped: No edges")
            continue
        
        # 3. 检查是否包含负边
        has_neg = any(is_label_negative(d.get('label', 0)) for _,_,d in lcc_G.edges(data=True))
        if not has_neg:
            # print(f"[Graph {idx}] Skipped: No negative edges found (Check Bitmask logic?)")
            continue
        
        print(f"[Graph {idx}] ENTERING MATCHING...")
        # count_run += 1
        
        metrics = compute_structural_confidence(lcc_G, bigG, MATCH_LIMIT, TIME_LIMIT, use_node_match=ENABLE_NODE_MATCH)
        
        if not metrics:
            # print(f"[Graph {idx}] Finished with NO metrics returned.")
            continue
        else:
            print(f"[Graph {idx}] SUCCESS! Returned metrics.")
        best_m = max(metrics, key=lambda x: x['confidence'])
        print(f"{idx:<4} | {best_m['original_label']:<4} | {best_m['confidence']:.4f}  | {best_m['support_negative']:<7} | {best_m['support_shape']:<9} | {best_m['label_dist']:<20} | {best_m['status']}")

        if best_m['support_shape'] > 0:
            for (u, v) in best_m['no_edge_pairs']:
                all_updates.append((u, v))

    print("UPDATE EDGES:",len(all_updates))
    if ENABLE_PROPAGATION and len(all_updates) > 0:
        unique_updates = list(set(all_updates))
        print(len(unique_updates))
    #     # ★★★ 传入新路径 ★★★
    #     apply_updates_to_csv_file(RAW_EDGE_FILE, UPDATE_EDGE_FILE, unique_updates)
```

```python
import os
import pandas as pd
import networkx as nx
from collections import Counter

# ================= 路径配置 =================
TARGET_FILE = "/home/yyyy/codework/GARplus/DiGress/outputs/2026-01-23/11-15-51-ppi_gar/generated_samples1.txt"
RAW_EDGE_FILE = "/home/yyyy/codework/GARplus/DiGress/DiGress-main/data/PPI/raw/protein_protein_with_type.csv"
RAW_NODE_FILE = "/home/yyyy/codework/GARplus/DiGress/DiGress-main/data/PPI/raw/protein.csv"

# ============== 你的当前弱标准编码 ==============
NODE_BIT_MAP = {'is_kinase': 0, 'is_disease_related': 1}
POS_EDGE_LABEL = 1
NEG_EDGE_LABEL = 2


def encode_node_feature(row_series):
    n_feature = 0
    keywords = str(row_series.get('Keywords', ''))
    families = str(row_series.get('Protein families', ''))
    disease = row_series.get('Involvement in disease', None)

    if 'Kinase' in keywords or 'ATP-binding' in keywords or 'Kinase' in families:
        n_feature |= (1 << NODE_BIT_MAP['is_kinase'])
    if pd.notna(disease) and str(disease).strip() != '':
        n_feature |= (1 << NODE_BIT_MAP['is_disease_related'])
    return n_feature


def get_edge_label(edge_type_str):
    s = str(edge_type_str).lower()
    is_negative = ('negative' in s) or ('added_negative' in s)
    return NEG_EDGE_LABEL if is_negative else POS_EDGE_LABEL


def is_label_negative(val):
    return int(val) == NEG_EDGE_LABEL


def load_csv_as_big_graph(edge_csv_path, node_csv_path):
    print("[Loader] Building BigGraph...")
    id_to_cat = {}

    df_meta = pd.read_csv(node_csv_path, low_memory=False)
    df_meta.columns = df_meta.columns.str.strip()

    # 尝试兼容不同列名
    id_col = None
    for cand in ['BioGRID ID', 'biogrid_id', 'BioGRID ID Interactor']:
        if cand in df_meta.columns:
            id_col = cand
            break
    if id_col is None:
        id_col = df_meta.columns[0]

    df_meta[id_col] = pd.to_numeric(df_meta[id_col], errors='coerce')
    df_meta = df_meta.dropna(subset=[id_col])

    for _, row in df_meta.iterrows():
        bid = str(int(row[id_col]))
        id_to_cat[bid] = encode_node_feature(row)

    BigG = nx.Graph()
    df_ppi = pd.read_csv(edge_csv_path, sep=',')
    if len(df_ppi.columns) < 5:
        df_ppi = pd.read_csv(edge_csv_path, sep='\t')
    df_ppi.columns = df_ppi.columns.str.strip()

    col_ua = 'Official Symbol Interactor A'
    col_ub = 'Official Symbol Interactor B'
    col_id_a = 'BioGRID ID Interactor A'
    col_id_b = 'BioGRID ID Interactor B'

    for _, row in df_ppi.iterrows():
        u_sym = str(row.get(col_ua, row.iloc[0]))
        v_sym = str(row.get(col_ub, row.iloc[1]))
        u_bid = str(row.get(col_id_a, '')).split('.')[0]
        v_bid = str(row.get(col_id_b, '')).split('.')[0]

        u_cat = id_to_cat.get(u_bid, 0)
        v_cat = id_to_cat.get(v_bid, 0)

        edge_type = row.get('type', 'positive')
        label = get_edge_label(edge_type)

        if not BigG.has_node(u_sym):
            BigG.add_node(u_sym, cat=u_cat)
        if not BigG.has_node(v_sym):
            BigG.add_node(v_sym, cat=v_cat)

        if BigG.has_edge(u_sym, v_sym):
            # 负边优先
            BigG[u_sym][v_sym]['label'] = max(BigG[u_sym][v_sym]['label'], label)
        else:
            BigG.add_edge(u_sym, v_sym, label=label)

    print(f"[Loader] BigGraph loaded. Nodes={BigG.number_of_nodes()}, Edges={BigG.number_of_edges()}")
    return BigG


def parse_graph_txt(filepath):
    graphs = []
    if not os.path.exists(filepath):
        raise FileNotFoundError(filepath)

    with open(filepath, "r") as f:
        lines = f.readlines()

    i = 0
    while i < len(lines):
        line = lines[i].strip()
        if line.startswith("N="):
            try:
                n = int(line.split("=")[1])

                i += 1
                while i < len(lines) and not lines[i].strip().startswith("X:"):
                    i += 1

                if i < len(lines):
                    i += 1  # skip X:
                x_list = list(map(int, lines[i].strip().split())) if i < len(lines) else []

                i += 1
                while i < len(lines) and not lines[i].strip().startswith("E:"):
                    i += 1

                edge_matrix = []
                if i < len(lines):
                    i += 1  # skip E:
                    for _ in range(n):
                        if i >= len(lines):
                            break
                        row = lines[i].strip().replace(',', ' ').split()
                        edge_matrix.append(list(map(int, row)))
                        i += 1

                graphs.append((x_list, edge_matrix))
            except Exception:
                i += 1
        else:
            i += 1

    return graphs


def generated_graph_stats(graphs_data):
    node_labels = Counter()
    edge_labels = Counter()
    total_edges = 0
    neg_edges = 0

    for x_list, edge_matrix in graphs_data:
        for x in x_list:
            node_labels[int(x)] += 1

        n = len(edge_matrix)
        for i in range(n):
            for j in range(i + 1, n):
                val = int(edge_matrix[i][j])
                if val != 0:
                    edge_labels[val] += 1
                    total_edges += 1
                    if is_label_negative(val):
                        neg_edges += 1

    return node_labels, edge_labels, total_edges, neg_edges


def big_graph_stats(bigG):
    node_labels = Counter()
    edge_labels = Counter()
    total_edges = 0
    neg_edges = 0

    for _, d in bigG.nodes(data=True):
        node_labels[int(d.get("cat", 0))] += 1

    for _, _, d in bigG.edges(data=True):
        val = int(d.get("label", POS_EDGE_LABEL))
        edge_labels[val] += 1
        total_edges += 1
        if is_label_negative(val):
            neg_edges += 1

    return node_labels, edge_labels, total_edges, neg_edges


def main():
    bigG = load_csv_as_big_graph(RAW_EDGE_FILE, RAW_NODE_FILE)
    graphs_data = parse_graph_txt(TARGET_FILE)

    print(f"[Info] Loaded generated graphs: {len(graphs_data)}")

    big_node_cnt, big_edge_cnt, big_total_e, big_neg_e = big_graph_stats(bigG)
    gen_node_cnt, gen_edge_cnt, gen_total_e, gen_neg_e = generated_graph_stats(graphs_data)

    big_node_space = set(big_node_cnt.keys())
    gen_node_space = set(gen_node_cnt.keys())

    big_edge_space = set(big_edge_cnt.keys())
    gen_edge_space = set(gen_edge_cnt.keys())

    print("\n===== Node Label Space =====")
    print("BigGraph node labels:", sorted(big_node_space))
    print("Generated node labels:", sorted(gen_node_space))
    print("Generated - BigGraph:", sorted(gen_node_space - big_node_space))
    print("BigGraph - Generated:", sorted(big_node_space - gen_node_space))

    print("\nTop node label counts in BigGraph:")
    print(big_node_cnt.most_common(10))
    print("Top node label counts in Generated:")
    print(gen_node_cnt.most_common(10))

    print("\n===== Edge Label Space =====")
    print("BigGraph edge labels:", sorted(big_edge_space))
    print("Generated edge labels:", sorted(gen_edge_space))
    print("Generated - BigGraph:", sorted(gen_edge_space - big_edge_space))
    print("BigGraph - Generated:", sorted(big_edge_space - gen_edge_space))

    print("\nTop edge label counts in BigGraph:")
    print(big_edge_cnt.most_common(10))
    print("Top edge label counts in Generated:")
    print(gen_edge_cnt.most_common(10))

    print("\n===== Negative / Non-negative Semantics =====")
    big_neg_ratio = big_neg_e / big_total_e if big_total_e > 0 else 0.0
    gen_neg_ratio = gen_neg_e / gen_total_e if gen_total_e > 0 else 0.0
    print(f"BigGraph total edges: {big_total_e}, negative edges: {big_neg_e}, neg ratio: {big_neg_ratio:.4f}")
    print(f"Generated total edges: {gen_total_e}, negative edges: {gen_neg_e}, neg ratio: {gen_neg_ratio:.4f}")

    print("\n===== Consistency Verdict =====")
    node_space_ok = len(gen_node_space - big_node_space) == 0
    edge_space_ok = len(gen_edge_space - big_edge_space) == 0

    print("Node label space consistent:", node_space_ok)
    print("Edge label space consistent:", edge_space_ok)

    if node_space_ok and edge_space_ok:
        print("PASS: 两个空间在标签集合层面是一致的。")
    else:
        print("FAIL: 两个空间在标签集合层面不一致。请重点看差集输出。")

    print("\n===== Sample Generated Graphs =====")
    for idx, (x_list, edge_matrix) in enumerate(graphs_data[:3]):
        print(f"\nGraph {idx}")
        print("  node labels:", x_list[:20])
        labels = []
        n = len(edge_matrix)
        for i in range(n):
            for j in range(i + 1, n):
                if edge_matrix[i][j] != 0:
                    labels.append(int(edge_matrix[i][j]))
        print("  edge labels:", labels[:20])


if __name__ == "__main__":
    main()
```

```python
import networkx as nx
import time
import signal
import matplotlib.pyplot as plt
from networkx.algorithms.isomorphism import GraphMatcher
import os
import traceback
import sys
from collections import Counter
import pandas as pd
import numpy as np

# ================= CONFIGURATION AREA =================
MATCH_LIMIT = 500
TIME_LIMIT = 5

# Input File (Generated Graph Samples)
TARGET_FILE = "/home/yyyy/codework/GARplus/DiGress/outputs/2026-03-11/12-13-57-ppi_gar/generated_samples1.txt"

# Original Data Files
RAW_EDGE_FILE = "/home/yyyy/codework/GARplus/DiGress/DiGress-main/data/PPI/raw/protein_protein_with_type.csv"
RAW_NODE_FILE = "/home/yyyy/codework/GARplus/DiGress/DiGress-main/data/PPI/raw/protein.csv"

# Output Update File
UPDATE_EDGE_FILE = "/home/yyyy/codework/GARplus/DiGress/DiGress-main/data/PPI/raw/extended_update/protein_protein_with_type_update.csv"

# Thresholds
ML_THRESHOLD = 0.5 

# BitMaps
NODE_BIT_MAP = {'is_kinase': 0, 'is_disease_related': 1}
EDGE_BIT_MAP = {"is_negative": 0}

# Feature Switches
# Recommended: False for initial topological matching to ensure hits
ENABLE_NODE_MATCH = True 
ENABLE_PROPAGATION = True  
# ======================================================

class TimeoutException(Exception): pass
def timeout_handler(signum, frame): raise TimeoutException()
if hasattr(signal, 'SIGALRM'):
    signal.signal(signal.SIGALRM, timeout_handler)

# ----------------------------------------------
# 1. Core Encoding Logic
# ----------------------------------------------
def encode_node_feature(row_series):
    """Encodes node features from protein.csv row."""
    n_feature = 0
    keywords = str(row_series.get('Keywords', ''))
    families = str(row_series.get('Protein families', ''))
    disease = row_series.get('Involvement in disease', None)
    
    if 'Kinase' in keywords or 'ATP-binding' in keywords or 'Kinase' in families:
        n_feature |= (1 << NODE_BIT_MAP['is_kinase'])
    if pd.notna(disease) and str(disease).strip() != '':
        n_feature |= (1 << NODE_BIT_MAP['is_disease_related'])
    return n_feature

def is_label_negative(val):
    """Checks if Bit 0 is 1 in the integer label."""
    if val == 0: return False
    mask = val - 1
    return (mask >> EDGE_BIT_MAP["is_negative"]) & 1

def get_big_graph_label(edge_type_str):
    """Parses CSV type: Positive -> 1, Negative -> 2"""
    s = str(edge_type_str).lower()
    if 'negative' in s or 'added_negative' in s: return 2
    return 1

def get_gar_edge_index(edge_type_str, score):
    """Helper to convert CSV type to label for loading."""
    is_neg = 1 if get_big_graph_label(edge_type_str) == 2 else 0
    return (is_neg << 0) + 1

# ----------------------------------------------
# 2. Matching Functions
# ----------------------------------------------
def node_match_fn(n1, n2):
    return n1.get('cat') == n2.get('cat')

def edge_match_fn(d1, d2):
    """
    d1: BigGraph Edge (Label 1 or 2)
    d2: GenGraph Edge (Bitmask Value)
    Matches if both are positive or both are negative.
    """
    l1 = d1.get('label', 0)
    l2 = d2.get('label', 0)
    return is_label_negative(l1) == is_label_negative(l2)

# ----------------------------------------------
# 3. BigGraph Loading Logic
# ----------------------------------------------
def load_csv_as_big_graph(edge_csv_path, node_csv_path):
    print(f"[Loader] Building BigGraph (Node Match: {ENABLE_NODE_MATCH})...")
    id_to_cat = {}
    
    # A. Load Node Metadata
    try:
        df_meta = pd.read_csv(node_csv_path, low_memory=False)
        df_meta.columns = df_meta.columns.str.strip()
        id_col = 'BioGRID ID' 
        if id_col not in df_meta.columns: id_col = df_meta.columns[0]
        df_meta[id_col] = pd.to_numeric(df_meta[id_col], errors='coerce').fillna(-1).astype(int).astype(str)
        
        for _, row in df_meta.iterrows():
            if row[id_col] == '-1': continue
            id_to_cat[row[id_col]] = encode_node_feature(row)
    except Exception as e:
        print(f"  - [Error] Meta: {e}"); return nx.Graph()

    # B. Load Edges
    BigG = nx.Graph()
    try:
        df_ppi = pd.read_csv(edge_csv_path, sep=',')
        if len(df_ppi.columns) < 5: df_ppi = pd.read_csv(edge_csv_path, sep='\t')
        df_ppi.columns = df_ppi.columns.str.strip()
        
        col_ua = 'Official Symbol Interactor A'
        col_ub = 'Official Symbol Interactor B'
        col_id_a = 'BioGRID ID Interactor A'
        col_id_b = 'BioGRID ID Interactor B'
        
        neg_count = 0
        for _, row in df_ppi.iterrows():
            # u_sym = str(row.get(col_ua, row[0])) 
            # v_sym = str(row.get(col_ub, row[1]))
            u_sym = str(row.get(col_ua, row.iloc[0]))
            v_sym = str(row.get(col_ub, row.iloc[1]))
            u_bid = str(row.get(col_id_a, '')).split('.')[0]
            v_bid = str(row.get(col_id_b, '')).split('.')[0]
            
            u_cat = id_to_cat.get(u_bid, 0)
            v_cat = id_to_cat.get(v_bid, 0)
            
            edge_type = row.get('type', 'positive')
            simple_label = get_big_graph_label(edge_type) # 1 or 2
            
            if simple_label == 2: neg_count += 1
            
            if not BigG.has_node(u_sym): BigG.add_node(u_sym, cat=u_cat)
            if not BigG.has_node(v_sym): BigG.add_node(v_sym, cat=v_cat)
            
            if BigG.has_edge(u_sym, v_sym):
                BigG[u_sym][v_sym]['label'] = max(BigG[u_sym][v_sym]['label'], simple_label)
            else:
                BigG.add_edge(u_sym, v_sym, label=simple_label)
        
        print(f"  - Loaded BigGraph Negative Edges: {neg_count}")
                
    except Exception as e:
        print(f"  - [Error] Edges: {e}"); return nx.Graph()

    print(f"BigGraph Loaded. Nodes: {BigG.number_of_nodes()}, Edges: {BigG.number_of_edges()}")
    return BigG

# ----------------------------------------------
# 4. Parsing & Structural Confidence
# ----------------------------------------------
def parse_graph_txt(filepath):
    graphs = []
    if not os.path.exists(filepath): return []
    with open(filepath, "r") as f: lines = f.readlines()
    i = 0
    while i < len(lines):
        line = lines[i].strip()
        if line.startswith("N="):
            try:
                n = int(line.split("=")[1])
                i += 1
                while i < len(lines) and not lines[i].strip().startswith("X:"): i += 1
                if i < len(lines): i += 1 
                x_list = list(map(int, lines[i].strip().split())) if i < len(lines) else []
                i += 1
                while i < len(lines) and not lines[i].strip().startswith("E:"): i += 1
                edge_matrix = []
                if i < len(lines):
                    i += 1
                    for _ in range(n):
                        if i >= len(lines): break
                        edge_matrix.append(list(map(int, lines[i].strip().replace(',', ' ').split())))
                        i += 1
                    graphs.append((x_list, edge_matrix))
            except: i += 1
        else: i += 1
    return graphs

def to_nx_graph(x_list, edge_matrix):
    n = len(edge_matrix)
    G = nx.Graph()
    for i in range(n): G.add_node(i, cat=x_list[i]) 
    for i in range(n):
        for j in range(i+1, n):
            if edge_matrix[i][j] != 0: G.add_edge(i, j, label=edge_matrix[i][j])
    return G

def compute_structural_confidence(subG, bigG, match_limit, time_limit, use_node_match=False):
    edges_to_test = list(subG.edges(data=True))
    target_candidates = [(u, v) for u, v, d in edges_to_test if is_label_negative(d.get('label', 0))]

    print(f"    [DEBUG] subG nodes={subG.number_of_nodes()}, edges={subG.number_of_edges()}, neg_targets={len(target_candidates)}")

    if not target_candidates:
        print("    [DEBUG] -> no negative target edges")
        return []

    nm_func = node_match_fn if use_node_match else None
    results = []

    for target_u, target_v in target_candidates:
        original_label = subG[target_u][target_v]['label']

        # 只删除目标负边，保留其余负边
        premiseG = subG.copy()
        premiseG.remove_edge(target_u, target_v)

        print(
            f"    [DEBUG] target=({target_u},{target_v}) "
            f"original_label={original_label} "
            f"premise_nodes={premiseG.number_of_nodes()} "
            f"premise_edges={premiseG.number_of_edges()} "
            f"removed_neg_edges=0"
        )

        if premiseG.number_of_edges() == 0:
            print("    [DEBUG] -> skip: premiseG empty after removing target edge")
            continue

        GM = GraphMatcher(bigG, premiseG, node_match=nm_func, edge_match=edge_match_fn)

        try:
            has_iso = GM.subgraph_is_isomorphic()
        except Exception as e:
            print(f"    [DEBUG] -> GraphMatcher init/check failed: {e}")
            continue

        if not has_iso:
            print("    [DEBUG] -> no subgraph isomorphism found")
            continue

        print("    [DEBUG] -> at least one isomorphism exists, entering iterator")

        supp_premise = 0
        supp_negative = 0
        label_dist = Counter()
        no_edge_pairs = []

        status = "Finished"
        try:
            if hasattr(signal, 'SIGALRM'):
                signal.alarm(time_limit)

            for mapping in GM.subgraph_isomorphisms_iter():
                real_u = mapping.get(target_u)
                real_v = mapping.get(target_v)

                if real_u is None or real_v is None:
                    continue

                supp_premise += 1

                real_label = 0
                if bigG.has_edge(real_u, real_v):
                    real_label = bigG[real_u][real_v].get('label', 0)

                label_dist[real_label] += 1

                if is_label_negative(real_label):
                    supp_negative += 1
                if real_label == 0:
                    no_edge_pairs.append((real_u, real_v))

                if supp_premise >= match_limit:
                    status = "Limit"
                    break

            if hasattr(signal, 'SIGALRM'):
                signal.alarm(0)

        except TimeoutException:
            status = "TimeOut"
            print("    [DEBUG] -> timeout")
        except Exception as e:
            status = "Error"
            print(f"    [DEBUG] -> iterator error: {e}")

        print(
            f"    [DEBUG] -> done: supp={supp_premise}, supp_neg={supp_negative}, "
            f"no_edge_pairs={len(no_edge_pairs)}, label_dist={dict(label_dist)}, status={status}"
        )

        if supp_premise > 0:
            confidence = supp_negative / supp_premise
            dist_str = str(dict(sorted(label_dist.items())))
            results.append({
                'target_edge': (target_u, target_v),
                'original_label': original_label,
                'confidence': confidence,
                'support_negative': supp_negative,
                'support_shape': supp_premise,
                'label_dist': dist_str,
                'no_edge_pairs': no_edge_pairs,
                'status': status
            })

    return results

# ==============================================================================
# 5. CSV Update Logic
# ==============================================================================
def apply_updates_to_csv_file(input_path, output_path, updates_list):
    if not updates_list: return
    print(f"\n[CSV Update] Reading original file: {input_path} ...")
    try:
        df = pd.read_csv(input_path, sep=',')
        if len(df.columns) < 5: df = pd.read_csv(input_path, sep='\t')
    except Exception as e: print(f"Error reading CSV: {e}"); return

    df.columns = df.columns.str.strip()
    col_a = 'Official Symbol Interactor A'
    col_b = 'Official Symbol Interactor B'
    col_type = 'type'
    if col_type not in df.columns: df[col_type] = 'positive' 
    
    print(f"[CSV Update] Processing {len(updates_list)} updates...")
    updated_count = 0
    new_rows = [] 
    
    # Remove duplicates
    updates_set = set(updates_list)
    
    # Pre-check existing edges to avoid slow lookups
    # (Simplified approach: iterate through updates)
    for u, v in updates_set:
        # Check if edge exists (A-B or B-A)
        mask = ((df[col_a] == u) & (df[col_b] == v)) | ((df[col_a] == v) & (df[col_b] == u))
        indices = df.index[mask].tolist()
        
        if len(indices) > 0:
            for idx in indices:
                old_val = df.at[idx, col_type]
                if old_val != 'added_negative':
                    df.at[idx, col_type] = 'added_negative'
                    updated_count += 1
        else:
            new_row = {col: np.nan for col in df.columns}
            new_row[col_a] = u
            new_row[col_b] = v
            new_row[col_type] = 'added_negative'
            if 'Score' in df.columns: new_row['Score'] = 0.0
            new_rows.append(new_row)

    if len(new_rows) > 0:
        print(f"[CSV Update] Appending {len(new_rows)} new rows...")
        df_new = pd.DataFrame(new_rows)
        df = pd.concat([df, df_new], ignore_index=True)

    print(f"\n[CSV Update] Summary: Modified {updated_count} existing rows, Appended {len(new_rows)} new rows.")
    print(f"[CSV Update] Saving to NEW file: {output_path} ...")
    
    os.makedirs(os.path.dirname(output_path), exist_ok=True)
    sep = '\t' if output_path.endswith('.tsv') else ','
    df.to_csv(output_path, index=False, sep=sep)
    print("Done! File saved successfully.")

# ----------------------------------------------
# 6. Main Execution
# ----------------------------------------------
if __name__ == "__main__":
    total_graphs = 0
    graphs_with_edges = 0
    graphs_with_neg = 0
    graphs_with_metrics = 0
    graphs_with_support_gt5 = 0
    graphs_with_no_edge_pairs = 0
    total_metric_items = 0
    # 1. Load Big Graph
    bigG = load_csv_as_big_graph(RAW_EDGE_FILE, RAW_NODE_FILE)
    if bigG.number_of_nodes() == 0: exit()

    # 2. Parse Generated Graphs
    graphs_data = parse_graph_txt(TARGET_FILE)
    print(f"Loaded {len(graphs_data)} generated graphs from txt.")
    
    print(f"\n{'ID':<4} | {'Type':<4} | {'Conf':<7} | {'SuppNeg':<7} | {'SuppShape':<9} | {'TruthDist':<20} | {'Status'}")
    print("-" * 90)
    
    all_updates = []
    
    for idx, (x_list, edge_mat) in enumerate(graphs_data):
        total_graphs += 1

        full_G = to_nx_graph(x_list, edge_mat)
        if full_G.number_of_nodes() == 0:
            continue

        comps = list(nx.connected_components(full_G))
        if not comps:
            continue

        lcc_G = full_G.subgraph(max(comps, key=len)).copy()
        if lcc_G.number_of_edges() < 1:
            continue
        graphs_with_edges += 1

        has_neg = any(is_label_negative(d.get('label', 0)) for _, _, d in lcc_G.edges(data=True))
        if not has_neg:
            continue
        graphs_with_neg += 1

        metrics = compute_structural_confidence(
            lcc_G, bigG, MATCH_LIMIT, TIME_LIMIT, use_node_match=ENABLE_NODE_MATCH
        )

        if not metrics:
            continue
        graphs_with_metrics += 1
        total_metric_items += len(metrics)

        best_m = max(metrics, key=lambda x: x['support_shape'])

        if best_m['support_shape'] > 5:
            graphs_with_support_gt5 += 1

        if len(best_m['no_edge_pairs']) > 0:
            graphs_with_no_edge_pairs += 1

        print(
            f"[Graph {idx}] support={best_m['support_shape']}, "
            f"neg={best_m['support_negative']}, "
            f"conf={best_m['confidence']:.4f}, "
            f"no_edge_pairs={len(best_m['no_edge_pairs'])}, "
            f"label_dist={best_m['label_dist']}, "
            f"status={best_m['status']}"
        )

    print("\n===== Debug Summary =====")
    print("total_graphs:", total_graphs)
    print("graphs_with_edges:", graphs_with_edges)
    print("graphs_with_neg:", graphs_with_neg)
    print("graphs_with_metrics:", graphs_with_metrics)
    print("total_metric_items:", total_metric_items)
    print("graphs_with_support_gt5:", graphs_with_support_gt5)
    print("graphs_with_no_edge_pairs:", graphs_with_no_edge_pairs)
```

```python
import os
import signal
from collections import Counter

import networkx as nx
import pandas as pd
import numpy as np
from networkx.algorithms.isomorphism import GraphMatcher

# ================= CONFIGURATION AREA =================
MATCH_LIMIT = 500
TIME_LIMIT = 10

TARGET_FILE = "/home/yyyy/codework/GARplus/DiGress/outputs/2026-03-11/12-13-57-ppi_gar/generated_samples1.txt"

RAW_EDGE_FILE = "/home/yyyy/codework/GARplus/DiGress/DiGress-main/data/PPI/raw/protein_protein_with_type.csv"
RAW_NODE_FILE = "/home/yyyy/codework/GARplus/DiGress/DiGress-main/data/PPI/raw/protein.csv"
UPDATE_EDGE_FILE = "/home/yyyy/codework/GARplus/DiGress/DiGress-main/data/PPI/raw/extended_update/protein_protein_with_type_update.csv"

ML_THRESHOLD = 0.3

# 与训练代码完全对齐
EDGE_BIT_MAP = {
    "is_negative": 0,
    "location_match_y": 1,
    "M_sim": 2,
    "physical_interaction": 3,
    "x_hub_degree": 4,
    "x_low_betweenness": 5,
    "Affinity_Capture_MS": 6,
}

ENABLE_NODE_MATCH = True
ENABLE_PROPAGATION = True

ONLY_FIRST_NEG_TARGET = True
MIN_PREMISE_EDGES = 1
MAX_PREMISE_EDGES = 100
# ======================================================


class TimeoutException(Exception):
    pass


def timeout_handler(signum, frame):
    raise TimeoutException()


if hasattr(signal, "SIGALRM"):
    signal.signal(signal.SIGALRM, timeout_handler)


# ----------------------------------------------
# 1. 编码逻辑：与训练代码对齐
# ----------------------------------------------
def map_loc_to_category(loc_str):
    s = str(loc_str).lower()
    if s == "nan" or s == "" or s == "-":
        return 9
    if "nucleus" in s or "nuclear" in s or "nucleoplasm" in s:
        return 0
    if "membrane" in s:
        return 2
    if "mitochondri" in s:
        return 4
    if "reticulum" in s:
        return 5
    if "golgi" in s:
        return 6
    if "lysosome" in s or "peroxisome" in s or "endosome" in s:
        return 7
    if "secreted" in s or "extracellular" in s:
        return 3
    if "cytoplasm" in s or "cytosol" in s:
        return 1
    return 8


def encode_node_feature(node_attr: dict) -> int:
    loc_raw = node_attr.get(
        "canon_Subcellular_location_CC",
        node_attr.get("Subcellular location [CC]", node_attr.get("location", "")),
    )
    return map_loc_to_category(loc_raw)


def _norm_str(v) -> str:
    if v is None:
        return ""
    if isinstance(v, float) and pd.isna(v):
        return ""
    return str(v).strip()


def encode_edge_feature(
    id_x,
    id_y,
    node_x_attr: dict,
    node_y_attr: dict,
    edge_row,
    global_stats: dict,
    sim_threshold: float = ML_THRESHOLD,
) -> int:
    e_feature = 0

    # A) location_match_y
    loc_x = _norm_str(
        node_x_attr.get(
            "canon_Subcellular_location_CC",
            node_x_attr.get("Subcellular location [CC]", node_x_attr.get("location", "")),
        )
    )
    loc_y = _norm_str(
        node_y_attr.get(
            "canon_Subcellular_location_CC",
            node_y_attr.get("Subcellular location [CC]", node_y_attr.get("location", "")),
        )
    )
    if loc_x and (loc_x == loc_y):
        e_feature |= (1 << EDGE_BIT_MAP["location_match_y"])

    # B) x_hub_degree
    try:
        deg_x = float(node_x_attr.get("degree", 0.0))
    except Exception:
        deg_x = 0.0
    q75_deg = float(global_stats.get("degree", {}).get("q75", float("inf")))
    if deg_x >= q75_deg:
        e_feature |= (1 << EDGE_BIT_MAP["x_hub_degree"])

    # C) x_low_betweenness
    try:
        bet_x = float(node_x_attr.get("betweenness_centrality", 0.0))
    except Exception:
        bet_x = 0.0
    q25_bet = float(global_stats.get("betweenness_centrality", {}).get("q25", float("-inf")))
    if bet_x <= q25_bet:
        e_feature |= (1 << EDGE_BIT_MAP["x_low_betweenness"])

    if edge_row:
        # D) M_sim
        try:
            ml_score = float(edge_row.get("ml_score", 0.0))
        except Exception:
            ml_score = 0.0
        if ml_score > sim_threshold:
            e_feature |= (1 << EDGE_BIT_MAP["M_sim"])

        # E1) Affinity Capture-MS
        exp_sys = _norm_str(edge_row.get("Experimental System", ""))
        if exp_sys == "Affinity Capture-MS":
            e_feature |= (1 << EDGE_BIT_MAP["Affinity_Capture_MS"])

        # E2) physical_interaction
        sys_type = str(edge_row.get("Experimental System Type", "")).lower()
        if "physical" in sys_type:
            e_feature |= (1 << EDGE_BIT_MAP["physical_interaction"])

        # E3) is_negative
        raw_type = str(edge_row.get("type", "positive")).lower()
        if ("negative" in raw_type) or (edge_row.get("is_negative") is True):
            e_feature |= (1 << EDGE_BIT_MAP["is_negative"])

    return int(e_feature)


def is_label_negative(val):
    """
    统一按训练时的标准：
    - Data/生成图里边标签存的是 edge_type = bitmask + 1
    - 如果是原始 bitmask（例如 bigG 里），也能兼容
    """
    val = int(val)
    if val == 0:
        return False

    # 如果是 txt / edge_attr 里出来的离散 edge_type
    mask = val - 1
    return ((mask >> EDGE_BIT_MAP["is_negative"]) & 1) == 1


# ----------------------------------------------
# 2. 匹配函数：与训练/验证逻辑对齐
# ----------------------------------------------
def node_match_fn(n1, n2):
    return (n1.get("feature_val") == n2.get("feature_val")) and (
        n1.get("deg", 0) <= n2.get("deg", 0)
    )


def edge_match_fn(d1, d2):
    # 严格按 bitmask label 一致匹配
    return int(d1.get("label", 0)) == int(d2.get("label", 0))


# ----------------------------------------------
# 3. 加载 BigGraph：与训练代码对齐
# ----------------------------------------------
def load_csv_as_big_graph(edge_csv_path, node_csv_path):
    print(f"[Loader] Building BigGraph (Node Match: {ENABLE_NODE_MATCH})...")
    id_to_attrs = {}

    try:
        df_meta = pd.read_csv(node_csv_path, low_memory=False)
        df_meta.columns = df_meta.columns.str.strip()

        # 与训练代码对齐
        id_col = "biogrid_id"
        if id_col not in df_meta.columns:
            id_col = "BioGRID ID" if "BioGRID ID" in df_meta.columns else df_meta.columns[0]

        df_meta[id_col] = pd.to_numeric(df_meta[id_col], errors="coerce")
        df_meta = df_meta.dropna(subset=[id_col])

        for _, row in df_meta.iterrows():
            bid = str(int(row[id_col]))
            id_to_attrs[bid] = row.to_dict()

    except Exception as e:
        print(f"  - [Error] Meta: {e}")
        return nx.Graph()

    BigG = nx.Graph()
    try:
        df_ppi = pd.read_csv(edge_csv_path, sep="," if "," in open(edge_csv_path).readline() else "\t")
        df_ppi.columns = df_ppi.columns.str.strip()

        # 第一遍：只加点加边，占位 label
        for _, row in df_ppi.iterrows():
            u_bid = str(row.get("BioGRID ID Interactor A", "")).split(".")[0]
            v_bid = str(row.get("BioGRID ID Interactor B", "")).split(".")[0]
            if not u_bid or not v_bid:
                continue

            u_attrs = id_to_attrs.get(u_bid, {})
            v_attrs = id_to_attrs.get(v_bid, {})

            u_enc = encode_node_feature(u_attrs)
            v_enc = encode_node_feature(v_attrs)

            BigG.add_node(u_bid, **u_attrs, feature_val=u_enc)
            BigG.add_node(v_bid, **v_attrs, feature_val=v_enc)

            edge_data = row.to_dict()
            BigG.add_edge(u_bid, v_bid, label=0, **edge_data)

        # 节点统计量
        deg_dict = dict(BigG.degree())
        nx.set_node_attributes(BigG, deg_dict, "degree")

        bet_dict = nx.betweenness_centrality(BigG, k=256, seed=42)
        nx.set_node_attributes(BigG, bet_dict, "betweenness_centrality")

        deg_vals = np.array(list(deg_dict.values()), dtype=float)
        bet_vals = np.array(list(bet_dict.values()), dtype=float)
        global_stats = {
            "degree": {"q75": float(np.quantile(deg_vals, 0.75))},
            "betweenness_centrality": {"q25": float(np.quantile(bet_vals, 0.25))},
        }

        # 第二遍：重算完整 bitmask label
        neg_count = 0
        for u, v, d in BigG.edges(data=True):
            ux = BigG.nodes[u]
            vy = BigG.nodes[v]
            bitmask = encode_edge_feature(
                id_x=u,
                id_y=v,
                node_x_attr=ux,
                node_y_attr=vy,
                edge_row=d,
                global_stats=global_stats,
            )
            d["label"] = int(bitmask)

            if bitmask & (1 << EDGE_BIT_MAP["is_negative"]):
                neg_count += 1

        for node in BigG.nodes():
            BigG.nodes[node]["deg"] = BigG.degree(node)

        print(f"  - Loaded BigGraph Negative Edges: {neg_count}")

    except Exception as e:
        print(f"  - [Error] Edges: {e}")
        return nx.Graph()

    print(f"BigGraph Loaded. Nodes: {BigG.number_of_nodes()}, Edges: {BigG.number_of_edges()}")
    return BigG


# ----------------------------------------------
# 4. 解析生成图
# ----------------------------------------------
def parse_graph_txt(filepath):
    graphs = []
    if not os.path.exists(filepath):
        return []

    with open(filepath, "r") as f:
        lines = f.readlines()

    i = 0
    while i < len(lines):
        line = lines[i].strip()
        if line.startswith("N="):
            try:
                n = int(line.split("=")[1])

                i += 1
                while i < len(lines) and not lines[i].strip().startswith("X:"):
                    i += 1
                if i < len(lines):
                    i += 1

                x_list = list(map(int, lines[i].strip().split())) if i < len(lines) else []

                i += 1
                while i < len(lines) and not lines[i].strip().startswith("E:"):
                    i += 1

                edge_matrix = []
                if i < len(lines):
                    i += 1
                    for _ in range(n):
                        if i >= len(lines):
                            break
                        edge_matrix.append(list(map(int, lines[i].strip().replace(",", " ").split())))
                        i += 1

                graphs.append((x_list, edge_matrix))
            except Exception:
                i += 1
        else:
            i += 1

    return graphs


def to_nx_graph(x_list, edge_matrix):
    """
    生成图里：
    - X 已经是 location category 0..9
    - E 里存的是 edge_type = bitmask + 1
    这里要还原成 bitmask，和 bigG 对齐
    """
    n = len(edge_matrix)
    G = nx.Graph()

    for i in range(n):
        G.add_node(i, feature_val=int(x_list[i]))

    for i in range(n):
        for j in range(i + 1, n):
            val = int(edge_matrix[i][j])
            if val != 0:
                bitmask = val - 1
                G.add_edge(i, j, label=bitmask)

    for node in G.nodes():
        G.nodes[node]["deg"] = G.degree(node)

    return G


# ----------------------------------------------
# 5. Structural Confidence
# ----------------------------------------------
def compute_structural_confidence(subG, bigG, match_limit, time_limit, use_node_match=False):
    edges_to_test = list(subG.edges(data=True))
    target_candidates = [
        (u, v) for u, v, d in edges_to_test
        if bool(int(d.get("label", 0)) & (1 << EDGE_BIT_MAP["is_negative"]))
    ]

    print(f"    [DEBUG] subG nodes={subG.number_of_nodes()}, edges={subG.number_of_edges()}, neg_targets={len(target_candidates)}")

    if not target_candidates:
        print("    [DEBUG] -> no negative target edges")
        return []

    if ONLY_FIRST_NEG_TARGET:
        target_candidates = target_candidates[:1]

    nm_func = node_match_fn if use_node_match else None
    results = []

    for target_u, target_v in target_candidates:
        original_label = subG[target_u][target_v]["label"]

        premiseG = subG.copy()
        premiseG.remove_edge(target_u, target_v)

        print(
            f"    [DEBUG] target=({target_u},{target_v}) "
            f"original_label={original_label} "
            f"premise_nodes={premiseG.number_of_nodes()} "
            f"premise_edges={premiseG.number_of_edges()} "
            f"removed_neg_edges=0"
        )

        if premiseG.number_of_edges() == 0:
            print("    [DEBUG] -> skip: premiseG empty after removing target edge")
            continue

        if premiseG.number_of_edges() < MIN_PREMISE_EDGES:
            print(f"    [DEBUG] -> skip: premise too weak (< {MIN_PREMISE_EDGES})")
            continue

        if premiseG.number_of_edges() > MAX_PREMISE_EDGES:
            print(f"    [DEBUG] -> skip: premise too dense (> {MAX_PREMISE_EDGES})")
            continue

        GM = GraphMatcher(bigG, premiseG, node_match=nm_func, edge_match=edge_match_fn)

        try:
            if hasattr(signal, "SIGALRM"):
                signal.alarm(time_limit)
            has_iso = GM.subgraph_is_isomorphic()
            if hasattr(signal, "SIGALRM"):
                signal.alarm(0)
        except TimeoutException:
            print("    [DEBUG] -> has_iso timeout")
            continue
        except Exception as e:
            if hasattr(signal, "SIGALRM"):
                signal.alarm(0)
            print(f"    [DEBUG] -> GraphMatcher init/check failed: {e}")
            continue

        if not has_iso:
            print("    [DEBUG] -> no subgraph isomorphism found")
            continue

        print("    [DEBUG] -> at least one isomorphism exists, entering iterator")

        supp_premise = 0
        supp_negative = 0
        label_dist = Counter()
        no_edge_pairs = []

        status = "Finished"
        try:
            if hasattr(signal, "SIGALRM"):
                signal.alarm(time_limit)

            for mapping in GM.subgraph_isomorphisms_iter():
                real_u = mapping.get(target_u)
                real_v = mapping.get(target_v)

                if real_u is None or real_v is None:
                    continue

                supp_premise += 1

                real_label = 0
                if bigG.has_edge(real_u, real_v):
                    real_label = int(bigG[real_u][real_v].get("label", 0))

                label_dist[real_label] += 1

                if bool(real_label & (1 << EDGE_BIT_MAP["is_negative"])):
                    supp_negative += 1
                if real_label == 0:
                    no_edge_pairs.append((real_u, real_v))

                if supp_premise >= match_limit:
                    status = "Limit"
                    break

            if hasattr(signal, "SIGALRM"):
                signal.alarm(0)

        except TimeoutException:
            status = "TimeOut"
            print("    [DEBUG] -> iterator timeout")
        except Exception as e:
            status = "Error"
            print(f"    [DEBUG] -> iterator error: {e}")

        print(
            f"    [DEBUG] -> done: supp={supp_premise}, supp_neg={supp_negative}, "
            f"no_edge_pairs={len(no_edge_pairs)}, label_dist={dict(label_dist)}, status={status}"
        )

        if supp_premise > 0:
            confidence = supp_negative / supp_premise
            dist_str = str(dict(sorted(label_dist.items())))
            results.append({
                "target_edge": (target_u, target_v),
                "original_label": original_label,
                "confidence": confidence,
                "support_negative": supp_negative,
                "support_shape": supp_premise,
                "label_dist": dist_str,
                "no_edge_pairs": no_edge_pairs,
                "status": status,
            })

    return results


# ----------------------------------------------
# 6. CSV Update Logic
# ----------------------------------------------
def apply_updates_to_csv_file(input_path, output_path, updates_list):
    if not updates_list:
        return

    print(f"\n[CSV Update] Reading original file: {input_path} ...")
    try:
        df = pd.read_csv(input_path, sep=",")
        if len(df.columns) < 5:
            df = pd.read_csv(input_path, sep="\t")
    except Exception as e:
        print(f"Error reading CSV: {e}")
        return

    df.columns = df.columns.str.strip()
    col_a = "Official Symbol Interactor A"
    col_b = "Official Symbol Interactor B"
    col_type = "type"
    if col_type not in df.columns:
        df[col_type] = "positive"

    print(f"[CSV Update] Processing {len(updates_list)} updates...")
    updated_count = 0
    new_rows = []

    updates_set = set(updates_list)

    for u, v in updates_set:
        mask = ((df[col_a] == u) & (df[col_b] == v)) | ((df[col_a] == v) & (df[col_b] == u))
        indices = df.index[mask].tolist()

        if len(indices) > 0:
            for idx in indices:
                old_val = df.at[idx, col_type]
                if old_val != "added_negative":
                    df.at[idx, col_type] = "added_negative"
                    updated_count += 1
        else:
            new_row = {col: np.nan for col in df.columns}
            new_row[col_a] = u
            new_row[col_b] = v
            new_row[col_type] = "added_negative"
            if "Score" in df.columns:
                new_row["Score"] = 0.0
            new_rows.append(new_row)

    if len(new_rows) > 0:
        print(f"[CSV Update] Appending {len(new_rows)} new rows...")
        df_new = pd.DataFrame(new_rows)
        df = pd.concat([df, df_new], ignore_index=True)

    print(f"\n[CSV Update] Summary: Modified {updated_count} existing rows, Appended {len(new_rows)} new rows.")
    print(f"[CSV Update] Saving to NEW file: {output_path} ...")

    os.makedirs(os.path.dirname(output_path), exist_ok=True)
    sep = "\t" if output_path.endswith(".tsv") else ","
    df.to_csv(output_path, index=False, sep=sep)
    print("Done! File saved successfully.")


# ----------------------------------------------
# 7. Main Execution
# ----------------------------------------------
if __name__ == "__main__":
    total_graphs = 0
    graphs_with_edges = 0
    graphs_with_neg = 0
    graphs_with_metrics = 0
    graphs_with_support_gt5 = 0
    graphs_with_no_edge_pairs = 0
    total_metric_items = 0

    bigG = load_csv_as_big_graph(RAW_EDGE_FILE, RAW_NODE_FILE)
    if bigG.number_of_nodes() == 0:
        raise SystemExit(1)

    graphs_data = parse_graph_txt(TARGET_FILE)
    print(f"Loaded {len(graphs_data)} generated graphs from txt.")

    print(f"\n{'ID':<4} | {'Type':<8} | {'Conf':<7} | {'SuppNeg':<7} | {'SuppShape':<9} | {'TruthDist':<20} | {'Status'}")
    print("-" * 100)

    all_updates = []

    for idx, (x_list, edge_mat) in enumerate(graphs_data):
        total_graphs += 1

        full_G = to_nx_graph(x_list, edge_mat)
        if full_G.number_of_nodes() == 0:
            continue

        comps = list(nx.connected_components(full_G))
        if not comps:
            continue

        lcc_G = full_G.subgraph(max(comps, key=len)).copy()
        if lcc_G.number_of_edges() < 1:
            continue
        graphs_with_edges += 1

        has_neg = any(
            bool(int(d.get("label", 0)) & (1 << EDGE_BIT_MAP["is_negative"]))
            for _, _, d in lcc_G.edges(data=True)
        )
        if not has_neg:
            continue
        graphs_with_neg += 1

        metrics = compute_structural_confidence(
            lcc_G, bigG, MATCH_LIMIT, TIME_LIMIT, use_node_match=ENABLE_NODE_MATCH
        )

        if not metrics:
            continue
        graphs_with_metrics += 1
        total_metric_items += len(metrics)

        best_m = max(metrics, key=lambda x: x["support_shape"])

        if best_m["support_shape"] > 5:
            graphs_with_support_gt5 += 1

        if len(best_m["no_edge_pairs"]) > 0:
            graphs_with_no_edge_pairs += 1

        print(
            f"[Graph {idx}] support={best_m['support_shape']}, "
            f"neg={best_m['support_negative']}, "
            f"conf={best_m['confidence']:.4f}, "
            f"no_edge_pairs={len(best_m['no_edge_pairs'])}, "
            f"label_dist={best_m['label_dist']}, "
            f"status={best_m['status']}"
        )

        if best_m["support_shape"] > 0:
            for (u, v) in best_m["no_edge_pairs"]:
                all_updates.append((u, v))

    print("\n===== Debug Summary =====")
    print("total_graphs:", total_graphs)
    print("graphs_with_edges:", graphs_with_edges)
    print("graphs_with_neg:", graphs_with_neg)
    print("graphs_with_metrics:", graphs_with_metrics)
    print("total_metric_items:", total_metric_items)
    print("graphs_with_support_gt5:", graphs_with_support_gt5)
    print("graphs_with_no_edge_pairs:", graphs_with_no_edge_pairs)
    print("raw_update_pairs:", len(all_updates))
    print("unique_update_pairs:", len(set(all_updates)))

    if ENABLE_PROPAGATION and len(all_updates) > 0:
        unique_updates = list(set(all_updates))
        # apply_updates_to_csv_file(RAW_EDGE_FILE, UPDATE_EDGE_FILE, unique_updates)
    else:
        print("\n[Result] No updates found.")
```

```python
import os
import pandas as pd
import networkx as nx
from collections import Counter

# ================= CONFIGURATION AREA =================
TARGET_FILE = "/home/yyyy/codework/GARplus/DiGress/outputs/2026-01-23/11-15-51-ppi_gar/generated_samples1.txt"
RAW_EDGE_FILE = "/home/yyyy/codework/GARplus/DiGress/DiGress-main/data/PPI/raw/protein_protein_with_type.csv"
RAW_NODE_FILE = "/home/yyyy/codework/GARplus/DiGress/DiGress-main/data/PPI/raw/protein.csv"

NODE_BIT_MAP = {'is_kinase': 0, 'is_disease_related': 1}

EDGE_BIT_MAP = {
    "is_negative": 0,
    "location_match_y": 1,
    "M_sim": 2,
    "physical_interaction": 3,
    "x_hub_degree": 4,
    "x_low_betweenness": 5,
    "Affinity_Capture_MS": 6
}

NUM_EDGE_BITS = len(EDGE_BIT_MAP)
MAX_GENERATED_EDGE_LABEL = (2 ** NUM_EDGE_BITS)  # bitmask max is 127, label max is 128

# ======================================================


def encode_node_feature(row_series):
    n_feature = 0
    keywords = str(row_series.get('Keywords', ''))
    families = str(row_series.get('Protein families', ''))
    disease = row_series.get('Involvement in disease', None)

    if 'Kinase' in keywords or 'ATP-binding' in keywords or 'Kinase' in families:
        n_feature |= (1 << NODE_BIT_MAP['is_kinase'])
    if pd.notna(disease) and str(disease).strip() != '':
        n_feature |= (1 << NODE_BIT_MAP['is_disease_related'])
    return n_feature


def is_label_negative(val):
    """
    Unified semantic decoder:
    - big graph: 1/2
    - generated graph: bitmask+1
    """
    val = int(val)
    if val == 0:
        return False
    mask = val - 1
    return ((mask >> EDGE_BIT_MAP["is_negative"]) & 1) == 1


def get_big_graph_label(edge_type_str):
    s = str(edge_type_str).lower()
    if 'negative' in s or 'added_negative' in s:
        return 2
    return 1


def load_csv_as_big_graph(edge_csv_path, node_csv_path):
    print("[Loader] Building BigGraph...")
    id_to_cat = {}

    df_meta = pd.read_csv(node_csv_path, low_memory=False)
    df_meta.columns = df_meta.columns.str.strip()

    id_col = 'BioGRID ID'
    if id_col not in df_meta.columns:
        id_col = df_meta.columns[0]

    df_meta[id_col] = pd.to_numeric(df_meta[id_col], errors='coerce').fillna(-1).astype(int).astype(str)

    for _, row in df_meta.iterrows():
        if row[id_col] == '-1':
            continue
        id_to_cat[row[id_col]] = encode_node_feature(row)

    BigG = nx.Graph()

    df_ppi = pd.read_csv(edge_csv_path, sep=',')
    if len(df_ppi.columns) < 5:
        df_ppi = pd.read_csv(edge_csv_path, sep='\t')
    df_ppi.columns = df_ppi.columns.str.strip()

    col_ua = 'Official Symbol Interactor A'
    col_ub = 'Official Symbol Interactor B'
    col_id_a = 'BioGRID ID Interactor A'
    col_id_b = 'BioGRID ID Interactor B'

    for _, row in df_ppi.iterrows():
        u_sym = str(row.get(col_ua, row.iloc[0]))
        v_sym = str(row.get(col_ub, row.iloc[1]))
        u_bid = str(row.get(col_id_a, '')).split('.')[0]
        v_bid = str(row.get(col_id_b, '')).split('.')[0]

        u_cat = id_to_cat.get(u_bid, 0)
        v_cat = id_to_cat.get(v_bid, 0)

        label = get_big_graph_label(row.get('type', 'positive'))

        if not BigG.has_node(u_sym):
            BigG.add_node(u_sym, cat=u_cat)
        if not BigG.has_node(v_sym):
            BigG.add_node(v_sym, cat=v_cat)

        if BigG.has_edge(u_sym, v_sym):
            BigG[u_sym][v_sym]['label'] = max(BigG[u_sym][v_sym]['label'], label)
        else:
            BigG.add_edge(u_sym, v_sym, label=label)

    print(f"BigGraph Loaded. Nodes: {BigG.number_of_nodes()}, Edges: {BigG.number_of_edges()}")
    return BigG


def parse_graph_txt(filepath):
    graphs = []
    if not os.path.exists(filepath):
        raise FileNotFoundError(filepath)

    with open(filepath, "r") as f:
        lines = f.readlines()

    i = 0
    while i < len(lines):
        line = lines[i].strip()
        if line.startswith("N="):
            try:
                n = int(line.split("=")[1])

                i += 1
                while i < len(lines) and not lines[i].strip().startswith("X:"):
                    i += 1
                if i < len(lines):
                    i += 1

                x_list = list(map(int, lines[i].strip().split())) if i < len(lines) else []

                i += 1
                while i < len(lines) and not lines[i].strip().startswith("E:"):
                    i += 1

                edge_matrix = []
                if i < len(lines):
                    i += 1
                    for _ in range(n):
                        if i >= len(lines):
                            break
                        row = lines[i].strip().replace(',', ' ').split()
                        edge_matrix.append(list(map(int, row)))
                        i += 1

                graphs.append((x_list, edge_matrix))
            except Exception:
                i += 1
        else:
            i += 1

    return graphs


def big_graph_stats(bigG):
    node_counter = Counter()
    edge_counter = Counter()
    neg_counter = Counter()

    for _, d in bigG.nodes(data=True):
        node_counter[int(d.get("cat", 0))] += 1

    for _, _, d in bigG.edges(data=True):
        val = int(d.get("label", 0))
        edge_counter[val] += 1
        neg_counter[is_label_negative(val)] += 1

    return node_counter, edge_counter, neg_counter


def generated_graph_stats(graphs_data):
    node_counter = Counter()
    edge_counter = Counter()
    neg_counter = Counter()

    for x_list, edge_matrix in graphs_data:
        for x in x_list:
            node_counter[int(x)] += 1

        n = len(edge_matrix)
        for i in range(n):
            for j in range(i + 1, n):
                val = int(edge_matrix[i][j])
                if val != 0:
                    edge_counter[val] += 1
                    neg_counter[is_label_negative(val)] += 1

    return node_counter, edge_counter, neg_counter


def check_generated_node_values(graphs_data):
    bad = []
    for gi, (x_list, _) in enumerate(graphs_data):
        for xi, x in enumerate(x_list):
            if int(x) not in {0, 1, 2, 3}:
                bad.append((gi, xi, int(x)))
                if len(bad) >= 20:
                    return bad
    return bad


def check_generated_edge_values(graphs_data):
    """
    For generated graphs, edge values are allowed to be:
      0 = no edge
      1..128 = bitmask+1
    """
    bad = []
    for gi, (_, edge_matrix) in enumerate(graphs_data):
        n = len(edge_matrix)
        for i in range(n):
            for j in range(i + 1, n):
                val = int(edge_matrix[i][j])
                if val < 0 or val > MAX_GENERATED_EDGE_LABEL + 1:
                    bad.append((gi, i, j, val))
                    if len(bad) >= 20:
                        return bad
    return bad


def main():
    bigG = load_csv_as_big_graph(RAW_EDGE_FILE, RAW_NODE_FILE)
    graphs_data = parse_graph_txt(TARGET_FILE)

    print(f"\n[Info] Loaded generated graphs: {len(graphs_data)}")

    big_node_cnt, big_edge_cnt, big_neg_cnt = big_graph_stats(bigG)
    gen_node_cnt, gen_edge_cnt, gen_neg_cnt = generated_graph_stats(graphs_data)

    big_node_space = set(big_node_cnt.keys())
    gen_node_space = set(gen_node_cnt.keys())

    print("\n===== 1. NODE LABEL SPACE =====")
    print("BigGraph node labels:", sorted(big_node_space))
    print("Generated node labels:", sorted(gen_node_space))
    print("Generated - BigGraph:", sorted(gen_node_space - big_node_space))
    print("BigGraph - Generated:", sorted(big_node_space - gen_node_space))
    print("Node space consistent:", len(gen_node_space - big_node_space) == 0)

    print("\n===== 2. RAW EDGE LABEL SPACE (informational only) =====")
    print("BigGraph edge labels:", sorted(big_edge_cnt.keys()))
    print("Generated edge labels:", sorted(gen_edge_cnt.keys())[:20], "...")

    print("\n===== 3. NEGATIVE-SEMANTIC SPACE =====")
    print("BigGraph neg/non-neg counts:", dict(big_neg_cnt))
    print("Generated neg/non-neg counts:", dict(gen_neg_cnt))

    print("\nBigGraph negative ratio:",
          big_neg_cnt[True] / (big_neg_cnt[True] + big_neg_cnt[False]))
    print("Generated negative ratio:",
          gen_neg_cnt[True] / (gen_neg_cnt[True] + gen_neg_cnt[False]) if (gen_neg_cnt[True] + gen_neg_cnt[False]) > 0 else 0.0)

    print("\n===== 4. LEGALITY CHECKS =====")
    bad_nodes = check_generated_node_values(graphs_data)
    bad_edges = check_generated_edge_values(graphs_data)

    print("Generated node values outside {0,1,2,3}:")
    print(bad_nodes if bad_nodes else "None")

    print("Generated edge values outside {0..128}:")
    print(bad_edges if bad_edges else "None")

    print("\n===== 5. FINAL VERDICT =====")
    node_ok = len(gen_node_space - big_node_space) == 0
    node_legal = len(bad_nodes) == 0
    edge_legal = len(bad_edges) == 0

    if node_ok and node_legal and edge_legal:
        print("PASS: 在“节点标签一致 + 只看负边语义”这个标准下，两个空间是兼容的。")
        print("解释：原始边标签整数空间不同，但通过 is_label_negative() 可映射到同一个二值语义空间。")
    else:
        print("FAIL: 两个空间在当前弱标准下仍有问题。")
        print("请重点检查：节点取值是否越界，或生成图边标签是否不是 bitmask+1 编码。")

    print("\n===== 6. SAMPLE GENERATED GRAPHS =====")
    for idx, (x_list, edge_matrix) in enumerate(graphs_data[:3]):
        labels = []
        n = len(edge_matrix)
        for i in range(n):
            for j in range(i + 1, n):
                if edge_matrix[i][j] != 0:
                    labels.append(int(edge_matrix[i][j]))

        neg_count = sum(1 for v in labels if is_label_negative(v))
        print(f"\nGraph {idx}")
        print("  node labels:", x_list[:20])
        print("  edge labels:", labels[:20])
        print("  neg-edge count:", neg_count)


if __name__ == "__main__":
    main()
```

```python
import os
from collections import Counter, defaultdict

import numpy as np

# ===== 路径配置 =====
TRAIN_TXT = "/home/yyyy/codework/GARplus/DiGress/outputs/2026-03-11/12-13-57-ppi_gar/ppi_train_ground_truth.txt"
GEN_TXT = "/home/yyyy/codework/GARplus/DiGress/outputs/2026-03-11/12-13-57-ppi_gar/generated_samples1.txt"

EDGE_BIT_MAP = {
    "is_negative": 0,
    "location_match_y": 1,
    "M_sim": 2,
    "physical_interaction": 3,
    "x_hub_degree": 4,
    "x_low_betweenness": 5,
    "Affinity_Capture_MS": 6,
}


def parse_graph_txt(filepath):
    graphs = []
    if not os.path.exists(filepath):
        raise FileNotFoundError(filepath)

    with open(filepath, "r", encoding="utf-8") as f:
        lines = f.readlines()

    i = 0
    while i < len(lines):
        line = lines[i].strip()
        if line.startswith("N="):
            try:
                n = int(line.split("=")[1])

                i += 1
                while i < len(lines) and not lines[i].strip().startswith("X:"):
                    i += 1
                if i < len(lines):
                    i += 1

                x_list = list(map(int, lines[i].strip().split())) if i < len(lines) else []

                i += 1
                while i < len(lines) and not lines[i].strip().startswith("E:"):
                    i += 1

                edge_matrix = []
                if i < len(lines):
                    i += 1
                    for _ in range(n):
                        if i >= len(lines):
                            break
                        row = list(map(int, lines[i].strip().replace(",", " ").split()))
                        edge_matrix.append(row)
                        i += 1

                graphs.append((x_list, np.array(edge_matrix, dtype=int)))
            except Exception:
                i += 1
        else:
            i += 1

    return graphs


def edge_type_to_bitmask(edge_type):
    if edge_type == 0:
        return None
    return int(edge_type) - 1


def is_negative_edge_type(edge_type):
    if edge_type == 0:
        return False
    bitmask = edge_type_to_bitmask(edge_type)
    return bool(bitmask & (1 << EDGE_BIT_MAP["is_negative"]))


def graph_signature_strict(x_list, E):
    """
    严格签名:
    - 节点标签序列
    - 上三角 edge_type 序列
    仅适用于固定节点编号下的精确比较
    """
    n = len(x_list)
    edge_vals = []
    for i in range(n):
        for j in range(i + 1, n):
            edge_vals.append(int(E[i, j]))
    return (tuple(map(int, x_list)), tuple(edge_vals))


def graph_signature_negative_semantic(x_list, E):
    """
    弱签名:
    - 节点标签序列
    - 上三角只保留负/非负/无边
    """
    n = len(x_list)
    edge_vals = []
    for i in range(n):
        for j in range(i + 1, n):
            et = int(E[i, j])
            if et == 0:
                edge_vals.append(0)
            else:
                edge_vals.append(2 if is_negative_edge_type(et) else 1)
    return (tuple(map(int, x_list)), tuple(edge_vals))


def collect_stats(graphs):
    stats = {
        "num_graphs": len(graphs),
        "node_labels": Counter(),
        "edge_types": Counter(),
        "graph_num_nodes": Counter(),
        "graph_num_edges": Counter(),
        "negative_edges": 0,
        "total_edges": 0,
    }

    for x_list, E in graphs:
        n = len(x_list)
        stats["graph_num_nodes"][n] += 1

        for x in x_list:
            stats["node_labels"][int(x)] += 1

        undirected_edges = 0
        for i in range(n):
            for j in range(i + 1, n):
                et = int(E[i, j])
                if et != 0:
                    undirected_edges += 1
                    stats["edge_types"][et] += 1
                    stats["total_edges"] += 1
                    if is_negative_edge_type(et):
                        stats["negative_edges"] += 1

        stats["graph_num_edges"][undirected_edges] += 1

    return stats


def print_stats(title, stats):
    print("\n" + "=" * 70)
    print(title)
    print("=" * 70)
    print(f"[Graphs] total = {stats['num_graphs']}")

    print("\n[Graph node count distribution]")
    for k, v in sorted(stats["graph_num_nodes"].items()):
        print(f"  N={k:<3} -> {v}")

    print("\n[Graph edge count distribution]")
    for k, v in sorted(stats["graph_num_edges"].items()):
        print(f"  E={k:<3} -> {v}")

    print("\n[Node label distribution]")
    total_nodes = sum(stats["node_labels"].values())
    for k, v in sorted(stats["node_labels"].items()):
        print(f"  label {k:<2} -> {v:<8} ({v / max(total_nodes,1):.2%})")

    print("\n[Top edge_type distribution]")
    for et, cnt in stats["edge_types"].most_common(20):
        print(f"  edge_type={et:<3} bitmask={et-1:<3} count={cnt}")

    neg_ratio = stats["negative_edges"] / max(stats["total_edges"], 1)
    print(f"\n[Negative edges] {stats['negative_edges']} / {stats['total_edges']} = {neg_ratio:.4%}")


def compare_graph_sets(train_graphs, gen_graphs):
    train_strict = Counter()
    train_neg_sem = Counter()

    for x_list, E in train_graphs:
        train_strict[graph_signature_strict(x_list, E)] += 1
        train_neg_sem[graph_signature_negative_semantic(x_list, E)] += 1

    strict_hit = 0
    neg_sem_hit = 0

    by_size = defaultdict(lambda: {"total": 0, "strict_hit": 0, "neg_sem_hit": 0})

    unmatched_examples = []

    for idx, (x_list, E) in enumerate(gen_graphs):
        n = len(x_list)
        by_size[n]["total"] += 1

        sig_s = graph_signature_strict(x_list, E)
        sig_n = graph_signature_negative_semantic(x_list, E)

        hit_s = sig_s in train_strict
        hit_n = sig_n in train_neg_sem

        if hit_s:
            strict_hit += 1
            by_size[n]["strict_hit"] += 1
        if hit_n:
            neg_sem_hit += 1
            by_size[n]["neg_sem_hit"] += 1

        if (not hit_s) and len(unmatched_examples) < 10:
            unmatched_examples.append({
                "idx": idx,
                "n": n,
                "strict_hit": hit_s,
                "neg_sem_hit": hit_n,
                "x": x_list,
            })

    print("\n" + "=" * 70)
    print("Generated vs Train Matching Report")
    print("=" * 70)
    print(f"Generated graphs total      : {len(gen_graphs)}")
    print(f"Strict exact hits           : {strict_hit} / {len(gen_graphs)} = {strict_hit / max(len(gen_graphs),1):.2%}")
    print(f"Negative-semantic exact hits: {neg_sem_hit} / {len(gen_graphs)} = {neg_sem_hit / max(len(gen_graphs),1):.2%}")

    print("\n[By graph size]")
    for n in sorted(by_size.keys()):
        row = by_size[n]
        total = row["total"]
        print(
            f"  N={n:<3} total={total:<3} "
            f"strict_hit={row['strict_hit']:<3} ({row['strict_hit']/max(total,1):.2%}) "
            f"neg_sem_hit={row['neg_sem_hit']:<3} ({row['neg_sem_hit']/max(total,1):.2%})"
        )

    print("\n[Sample unmatched generated graphs]")
    for ex in unmatched_examples:
        print(
            f"  idx={ex['idx']:<3} N={ex['n']:<3} "
            f"strict_hit={ex['strict_hit']} neg_sem_hit={ex['neg_sem_hit']} "
            f"x={ex['x'][:20]}"
        )


def main():
    print(f"[Load] train txt: {TRAIN_TXT}")
    train_graphs = parse_graph_txt(TRAIN_TXT)

    print(f"[Load] generated txt: {GEN_TXT}")
    gen_graphs = parse_graph_txt(GEN_TXT)

    train_stats = collect_stats(train_graphs)
    gen_stats = collect_stats(gen_graphs)

    print_stats("Train TXT Statistics", train_stats)
    print_stats("Generated TXT Statistics", gen_stats)

    compare_graph_sets(train_graphs, gen_graphs)


if __name__ == "__main__":
    main()
```

```python
import torch
import os
import sys
import numpy as np
from tqdm import tqdm

# 确保能导入 src (假设脚本在项目根目录)
sys.path.append(os.getcwd())

# 尝试导入数据集类 (请根据你的实际目录结构调整)
try:
    from src.datasets.ppi_dataset import PPIGraphDataset
except ImportError:
    # 如果找不到，尝试添加上一级目录
    sys.path.append(os.path.join(os.getcwd(), '..'))
    from src.datasets.ppi_dataset import PPIGraphDataset

# ==========================================
# 1. 定义特征映射 (用于统计诊断)
# ==========================================
# 必须与你训练时使用的映射完全一致
NODE_BIT_MAP = {
    'is_kinase': 0,
    'is_disease_related': 1
}

EDGE_BIT_MAP = {
    "is_negative": 0,           # Bit 0
    "location_match_y": 1,      # Bit 1
    "M_sim": 2,                 # Bit 2
    "physical_interaction": 3,  # Bit 3
    "x_hub_degree": 4,          # Bit 4
    "x_low_betweenness": 5,     # Bit 5
    "Affinity_Capture_MS": 6,   # Bit 6
}

# ==========================================
# 2. 辅助函数
# ==========================================

def get_int_labels(tensor, is_edge=False):
    """
    智能将 Tensor 转换为整数标签 (0, 1, 2...)
    兼容 One-Hot (float) 和 Raw Labels (long)
    """
    if tensor is None or tensor.numel() == 0:
        return None
    
    # 将 tensor 移到 CPU 并转为 numpy
    t = tensor.cpu()
    
    # 情况 A: 已经是整数标签 [N] 或 [N, 1]
    if t.dtype in [torch.long, torch.int, torch.int32, torch.int64]:
        return t.numpy().flatten()
    
    # 情况 B: One-Hot 编码 [N, C] (通常是 float)
    if t.dim() > 1 and t.shape[-1] > 1:
        return t.argmax(dim=-1).numpy()
    
    # 情况 C: [N, 1] 的 float，强制转 int
    return t.int().numpy().flatten()

def tensor_to_dense_matrix(edge_index, edge_attr, num_nodes):
    """
    将稀疏边列表转换为密集邻接矩阵 E
    """
    E = np.zeros((num_nodes, num_nodes), dtype=int)
    
    if edge_index is not None and edge_index.numel() > 0:
        src = edge_index[0].cpu().numpy()
        dst = edge_index[1].cpu().numpy()
        
        # 获取边的类别索引 (0-128)
        # 如果 edge_attr 是空的 (无特征图)，则默认为 1 (也就是存在边，但无特殊属性)
        if edge_attr is not None and edge_attr.numel() > 0:
            attr_vals = get_int_labels(edge_attr, is_edge=True)
        else:
            attr_vals = np.ones(len(src), dtype=int) # 默认值为1

        for i in range(len(src)):
            u, v = src[i], dst[i]
            if u < num_nodes and v < num_nodes:
                E[u, v] = attr_vals[i]
            
    return E

# ==========================================
# 3. 主导出函数
# ==========================================

def export_dataset(split='train', output_file='train_graphs_ground_truth.txt'):
    # 请根据你的实际路径修改
    DATA_ROOT = "/home/yyyy/codework/GARplus/DiGress/DiGress-main/data/PPI"
    
    output_path = os.path.join("/home/yyyy/codework/GARplus/DiGress/outputs/2025-12-15/19-53-11-ppi_gar", output_file)
    # 确保输出目录存在
    os.makedirs(os.path.dirname(output_path), exist_ok=True)

    print(f"Loading {split} dataset from {DATA_ROOT}...")
    try:
        dataset = PPIGraphDataset(root=DATA_ROOT, split=split)
    except Exception as e:
        print(f"Error loading dataset: {e}")
        return

    print(f"Found {len(dataset)} graphs. Exporting to {output_path}...")
    
    # --- 统计计数器 ---
    stats = {
        "total_edges": 0,
        "edge_bits": {k: 0 for k in EDGE_BIT_MAP.keys()},
        "total_nodes": 0,
        "node_bits": {k: 0 for k in NODE_BIT_MAP.keys()}
    }

    with open(output_path, 'w', encoding='utf-8') as f:
        for i, data in tqdm(enumerate(dataset), total=len(dataset)):
            num_nodes = data.num_nodes
            stats["total_nodes"] += num_nodes
            
            # --- 1. 处理节点特征 X ---
            # 获取整数标签 (0-3)
            x_arr = get_int_labels(data.x)
            if(x_arr is None or len(x_arr) != num_nodes):
                 # 容错：如果读取失败，填默认值 0
                 x_arr = np.zeros(num_nodes, dtype=int)
            
            # 统计节点 Bit
            for val in x_arr:
                for k, bit in NODE_BIT_MAP.items():
                    if (val >> bit) & 1:
                        stats["node_bits"][k] += 1

            # --- 2. 处理边特征 E ---
            # 获取密集矩阵
            e_arr = tensor_to_dense_matrix(data.edge_index, data.edge_attr, num_nodes)
            
            # 统计边 Bit
            rows, cols = e_arr.shape
            for u in range(rows):
                for v in range(u + 1, cols):
                    val = int(e_arr[u, v])
                    if val == 0: continue # 无边
                    
                    stats["total_edges"] += 1
                    # 还原 Bitmask: 类别值 = mask + 1 -> mask = val - 1
                    mask = val - 1
                    
                    for k, bit in EDGE_BIT_MAP.items():
                        if (mask >> bit) & 1:
                            stats["edge_bits"][k] += 1

            # --- 3. 写入文件 ---
            f.write(f"N={num_nodes}\n")
            f.write("X:\n")
            f.write(" ".join(map(str, x_arr)) + "\n")
            f.write("E:\n")
            for row in e_arr:
                f.write(" ".join(map(str, row)) + "\n")
            f.write("\n")
            
    # ==========================================
    # 4. 打印诊断报告 (最关键的部分)
    # ==========================================
    print("\n" + "="*60)
    print(f"Dataset Diagnosis Report ({split})")
    print("="*60)
    
    # 节点统计
    print(f"\n[Nodes] Total: {stats['total_nodes']}")
    if stats['total_nodes'] > 0:
        for k, v in stats['node_bits'].items():
            ratio = v / stats['total_nodes']
            print(f"  - {k:<20}: {v:<8} ({ratio:.2%})")
            
    # 边统计
    print(f"\n[Edges] Total: {stats['total_edges']}")
    if stats['total_edges'] > 0:
        for k, v in stats['edge_bits'].items():
            ratio = v / stats['total_edges']
            print(f"  - {k:<20}: {v:<8} ({ratio:.2%})")
            
    print("\n" + "="*60)
    print(f"Export finished: {output_path}")

if __name__ == "__main__":
    export_dataset(split='train')
```

```python
import os
import sys
import numpy as np
from tqdm import tqdm

# ==========================================
# 1. 定义特征映射 (核心逻辑)
# ==========================================
NODE_BIT_MAP = {
    'is_kinase': 0,
    'is_disease_related': 1
}

EDGE_BIT_MAP = {
    "is_negative": 0,           # Bit 0
    "location_match_y": 1,      # Bit 1
    "M_sim": 2,                 # Bit 2
    "physical_interaction": 3,  # Bit 3
    "x_hub_degree": 4,          # Bit 4
    "x_low_betweenness": 5,     # Bit 5
    "Affinity_Capture_MS": 6,   # Bit 6
}

# ==========================================
# 2. TXT 解析器
# ==========================================
def parse_and_diagnose(file_path):
    if not os.path.exists(file_path):
        print(f"❌ Error: 文件不存在 -> {file_path}")
        return

    print(f"Reading and analyzing: {file_path} ...")
    
    # --- 统计计数器 ---
    stats = {
        "total_graphs": 0,
        "total_nodes": 0,
        "total_edges": 0,
        "node_bits": {k: 0 for k in NODE_BIT_MAP.keys()},
        "edge_bits": {k: 0 for k in EDGE_BIT_MAP.keys()}
    }

    # 读取所有行
    with open(file_path, 'r', encoding='utf-8') as f:
        lines = f.readlines()

    # 使用指针遍历文件
    i = 0
    total_lines = len(lines)
    
    # 简单的进度条封装
    pbar = tqdm(total=total_lines, desc="Scanning Graphs")

    while i < total_lines:
        line = lines[i].strip()
        
        # 跳过空行
        if not line:
            i += 1
            pbar.update(1)
            continue

        # 1. 识别图的开始 (通常是 N=...)
        # 但我们主要依靠识别 X: 和 E:
        if line.startswith("X:"):
            # 下一行是节点特征
            if i + 1 >= total_lines: break
            
            try:
                x_line = lines[i+1].strip()
                # 解析 X 数组
                x_arr = np.array([int(v) for v in x_line.split()])
                num_nodes = len(x_arr)
                
                # 统计节点特征
                stats["total_graphs"] += 1
                stats["total_nodes"] += num_nodes
                
                for val in x_arr:
                    for k, bit in NODE_BIT_MAP.items():
                        if (val >> bit) & 1:
                            stats["node_bits"][k] += 1
                
                i += 2 # 跳过 "X:" 和数据行
                pbar.update(2)
            except ValueError:
                print(f"Warning: Failed to parse X at line {i+1}")
                i += 1
                pbar.update(1)
                continue

            # 2. 紧接着应该是 E:
            if i < total_lines and lines[i].strip().startswith("E:"):
                i += 1 # 跳过 "E:"
                pbar.update(1)
                
                # 解析 E 矩阵 (接下来的 num_nodes 行)
                e_matrix = []
                for _ in range(num_nodes):
                    if i >= total_lines: break
                    row_vals = [int(v) for v in lines[i].strip().split()]
                    e_matrix.append(row_vals)
                    i += 1
                    pbar.update(1)
                
                e_arr = np.array(e_matrix)
                
                # 统计边特征
                # 注意：只统计上三角矩阵 (避免无向图重复统计)
                rows, cols = e_arr.shape
                for u in range(rows):
                    for v in range(u + 1, cols):
                        val = int(e_arr[u, v])
                        if val == 0: continue # 无边
                        
                        stats["total_edges"] += 1
                        
                        # ★★★ 核心解码逻辑 ★★★
                        # 存储值 = mask + 1  ==>  mask = 存储值 - 1
                        mask = val - 1
                        
                        for k, bit in EDGE_BIT_MAP.items():
                            if (mask >> bit) & 1:
                                stats["edge_bits"][k] += 1
            else:
                print(f"Warning: Expected 'E:' after 'X:' at line {i}")
        
        else:
            # 可能是 N=... 或者其他行，跳过
            i += 1
            pbar.update(1)

    pbar.close()
    
    # ==========================================
    # 3. 打印诊断报告
    # ==========================================
    print_report(stats, file_path)

def print_report(stats, file_name):
    print("\n" + "="*60)
    print(f"Diagnosis Report for: {os.path.basename(file_name)}")
    print("="*60)
    
    print(f"Total Graphs Scanned: {stats['total_graphs']}")

    # --- 节点统计 ---
    print(f"\n[Nodes] Total: {stats['total_nodes']}")
    if stats['total_nodes'] > 0:
        print(f"{'Feature':<25} | {'Count':<8} | {'Ratio':<8}")
        print("-" * 45)
        for k, v in stats['node_bits'].items():
            ratio = v / stats['total_nodes']
            print(f"{k:<25} | {v:<8} | {ratio:.2%}")
            
    # --- 边统计 ---
    print(f"\n[Edges] Total: {stats['total_edges']}")
    if stats['total_edges'] > 0:
        print(f"{'Feature':<25} | {'Count':<8} | {'Ratio':<8}")
        print("-" * 45)
        for k, v in stats['edge_bits'].items():
            ratio = v / stats['total_edges']
            print(f"{k:<25} | {v:<8} | {ratio:.2%}")
    else:
        print("⚠️ No edges found! (Matrix might be all zeros)")

    print("\n" + "="*60)

# ==========================================
# 4. 运行配置
# ==========================================
if __name__ == "__main__":
    # 在这里填入你要分析的 TXT 文件路径
    # 例如：生成的样本，或者导出的 Ground Truth
    # TARGET_FILE = "/home/yyyy/codework/GARplus/DiGress/outputs/2025-12-15/19-53-11-ppi_gar/train_graphs_ground_truth.txt"
    TARGET_FILE = "/home/yyyy/codework/GARplus/DiGress/outputs/2026-01-23/11-15-51-ppi_gar/generated_samples1.txt"
    
    parse_and_diagnose(TARGET_FILE)
```
